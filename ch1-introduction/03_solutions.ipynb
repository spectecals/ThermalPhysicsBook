## Solutions

1. What is the mass of 3 moles of carbon dioxide (CO<sub>2</sub>)?
```{admonition} Solution
:class: dropdown

CO<sub>2</sub> is composed of C and O with atomic numbers $Z = 12, 16$. 

$$
\text{total molar mass} \mult \text{number of moles} = \text{mass of gas in grams} \\
(12 + 16\mult 2)\mult 3 = 44 \mult 3 = 132\ \text{g}
$$

```

2. A typical bacterium has a mass of $10^{-12}$ g. Calculate the mass of a mole of bacteria in units of elephant-masses (elephants have a mass ~5000 kg).
```{admonition} Solution
:class: dropdown

Here we require Avagadro's number, $6.022 \cdot 10^{23}$, and it will help to
write the elephant masses in the form of $5 \cdot 10^{3}\ \text{kg} = 5 \cdot 10^{6}\ \text{g}$.

$$
\begin{aligned}
\text{single bacterium mass} \times \text{Avagadro's number} / \text{elephant mass}
&= \text{elephants composed of bacteria} \\
10^{-12}\ \text{g} \times 6.022 \cdot 10^{23}\ \text{mol}^{-1} \mult \frac{1}{5} \cdot 10^{6}\ \text{g}^{-1} &= 2 \times 6.022 \cdot 10^{(23 - 19)} \\
= 12.044 \cdot 10^{4} &= 120,000\text{ elephants}
\end{aligned}
$$
      
```

3. Are there more water *molecules* in your body (assuming you are 100% water) or more water *droplets* in all the oceans in the world? There is approximately $10^{21}$ kg of water in all of the world's oceans. Use an estimate of the size of a typical droplet of water.
```{admonition} Solution
:class: dropdown

Average weight of a European is $\approx 70\ \text{kg}$. H<sub>2</sub>O has a
molar mass of $(2 + 16)$ g mol<sup>-1</sup>. To solve this question, we'll need
to calculate the number of molecules in a human, and then how many water droplets
in the ocean. We'll start by first determining how many
moles of water constitute a mass of 70 kg, and then how many molecules total.

$$
\begin{aligned}
70,000\ \text{g} / 18\ \text{g mol}^{-1} &\approx 3,900\ \text{mol} \\
3,900\ \text{mol}\times 6.022 \cdot 10^{23}\ \text{molecules per mol} &= 2.35 \cdot 10^{27}\ \text{molecules}
\end{aligned}
$$

Now to calculate how many droplets are in the ocean. Let's use a reasonable
assumption for the size of a droplet based on raindrops which range from 
0.5â€“5 mm typically, so we'll go with 2 mm diameter droplets.

$$
\text{volume}=\frac{4}{3}\pi r^2=\frac{4}{3}\pi (1 \cdot 10^{-3})^3\ \text{m}^3 &\approx 10^{-9}\ \text{m}^3 \\
$$

Given that water has density 1,000 kg m<sup>-3 </sup>,
we can directly divide our total mass of the oceans by this volume after converting to kg.

$$
\begin{aligned}
1\ \text{m}^3\ \text{water} \times 1,000\ \text{kg m}^{-3} &\equiv 1,000\ \text{kg water} \\
10^{-9}\ \text{m}^3\ \text{(droplet)}&= 10^{-6}\ \text{kg} \\
10^{21}\ \text{kg (ocean)} \times 10^{6} &= 10^{27} < 2.35 \cdot 10^{27}
\end{aligned}
$$

Depending on your approximation of the droplet size, it could go either way! If
you used 0.5 mm diameter droplets, you'd have $8 \cdot 10^{27} > 2.35 \cdot 10^{27}$.
```

```{admonition} 2. $r=2\cdot 10^{23},\ r=10^{23}$?
:class: dropdown

Using $^nC_R$ here is considerably more difficult:

$$
\frac{2\cdot 10^{23}!}{10^{23}!(2\cdot 10^{23}-10^{23})!} = \frac{2\cdot 10^{23}!}{10^{23}!\cdot 10^{23}!} = \frac{2\cdot r!}{r!\cdot r!}.
$$

We can approximate the numerator and denominator via Stirling's:

$$
\begin{aligned}
\ln 2r! &\approx 2r \ln 2r - 2r \\
\ln (r! \cdot r!) &= 2(\ln r!) \approx 2(r \ln r - r)\\
&= 2r \ln r - 2r
\end{aligned}
$$

Dividing the two,

$$
\begin{aligned}
\ln\frac{2r!}{r! \cdot r!} &= 2r \ln 2r - 2r - (2r \ln r - 2r) \\
&= 2r(\ln 2r - \ln r) \\
&= 2r \ln 2 \\
&= 2 \cdot 10^{23} \ln 2 \\
&= 1.39 \cdot 10^{23}
\end{aligned}
$$

Finally, we need to exponentiate our result, which gives us the number of arrangements:

$$
\text{(base e)}\ e^{1.39 \cdot 10^{23}} = 10^{1.39 \cdot 10^{23} / \ln 10} \approx 10^{6 \cdot 10^{22}}\ \text{(base 10)}.
$$

Either form is acceptable, base 10 just shows us the number of digits are in our answer.
```

```{admonition} 2. Give three examples of these variables.
:class: dropdown
Pressure, volume, and temperature.
```
```{admonition} 3. After the gas is compressed, what condition must be satisfied for these variables to be well defined again?
:class: dropdown
The gas must thermalise, i.e., reach thermal equilibrium, such that a single value for $T$ or $P$ can describe the full system.
```
```{admonition} 4. How do the thermodynamic variables change when the gas is compressed?
:class: dropdown
The volume decreases. With essentially no time for heat to leave, the temperature and pressure increase.
```
```{admonition} 5. Why can these variables not be used to describe the gas during the compression itself?
:class: dropdown
The variables are not stable during this timeframe, i.e., the elevated temperature will dissipate to the surroundings immediately after compression, and the temperature across the gas will vary. In the centre, the temperature will likely be higher than the gas contacting the walls which are transferring heat out of the system. As a result, we cannot assign a singular value for $T$, and therefore the pressure will also vary across the system too.
```
