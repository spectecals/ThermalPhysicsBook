## Supercooled Liquids Project

Your task is to conduct the supercooled liquid experiment at home and report your results. [See here for instructions!](https://www.youtube.com/watch?v=ph8xusY3GTM&t=30s)

You should film your results and then prepare a presentation about the physics and your results. The assessment will take the form of a 10-minute PowerPoint presentation that you will narrate.

Previously, students have found using Evian water is helpful. You can also try sparkling water, boiled water, or fizzy drinks.

You will be marked on the following:

- Overall quality and accuracy of the scientfici content (25%)
- Quality of the presentation material (25%)
- Quality of the delviery of the content (25%)
- Added value. Have you provided extra insight or done additional research around the topic? (25%)
