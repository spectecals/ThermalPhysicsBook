## Question 2 - Gas Compression

Calculate the work done by the gas and the heat flow into the gas.

## Question 6 - Heat Capacities and the Gas Constant (TP 11.2)

Starting from the definitions of $C_V$, $C_p$ and $\gamma$, show that, for an ideal gas:

$$
\frac{R}{C_V} = \gamma - 1
$$

and

$$
\frac{R}{C_p} = \frac{\gamma - 1}{\gamma},
$$

where $C_V$ and $C_p$ are the heat capacities per mole.

## Question 9 - Path Independence (TP 11.3)

Consider the differential:

$$
dz = 2xy \ dx + (x^2 + 2y) \ dy
$$

Evaluate the integral $\int^{(x_2, y_2)}_{(x_1, y_1)}dz$ along the paths consisting of the straight line segments:

(a) $(x_1, y_1) \rightarrow (x_2, y_1)$ and then $(x_2, y_1) \rightarrow (x_2, y_2)$.

(b) $(x_1, y_1) \rightarrow (x_1, y_2)$ and then $(x_1, y_2) \rightarrow (x_2, y_2)$.

Is $dz$ an exact differential? (Does the integral depend only on its start and end points?)
