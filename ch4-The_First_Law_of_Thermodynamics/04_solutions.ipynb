## Question 1 - Cloud formation

As air rises over a mountain, it moves into a region of lower atmospheric pressure. The air parcel expands as it rises and, in doing so, does work on its surroundings. During this process, the air is not in meaningful thermal contact with the mountain or its surroundings, so no heat is being transferred as it ascends the mountain. Therefore, we can treat this circumstance as an adiabatic process ($Q = 0$).

From the first law of thermodynamics:

$$
\Delta U = Q + W_{on}.
$$

Since the gas is expanding, work is done *by* the gas and $W_{on} < 0$. This leads to a decrease in internal energy and hence a decrease in the vapour's temperature. As the air cools, the vapour pressure of the water decreases. Once the vapour pressure drops below the partial pressure threshold, the water vapour must condense into liquid droplets. These droplets form clouds, which is why cloud formation often occurs near the tops of mountains.

## Question 2 - Gas Compression

#### Case 1: Isothermal Compression ($T = \text{constant}$)

For an ideal gas, internal energy depends only on its temperature. So if $T$ is constant then $\Delta U = 0$. Since the gas is being compressed, work is being done *on* the gas, so our expression for this work is:

$$
W_{on} = - \int_{V_1}^{V_2} p dV.
$$

Noting that $p = \tfrac{nRT}{V}$ and that $T$ is a constant, we can write:

$$
W_{on} = - \int_{V_1}^{V_2} \frac{nRT}{V} dV = -nRT \int_{V_1}^{V_2} \frac{1}{V} dV = - nRT \ln{\left(\frac{V_2}{V_1}\right)}.
$$

Since $V_2 < V_1$, $\ln(V_2/V_1) < 0$ and hence the minus signs cancel out and ensure $W_{on} > 0$, and equivalently, $W_{by}$ is assured to be < 0.

#### Case 2: Adiabatic Compression ($Q = 0$)

Because the process is adiabatic, we have:

$$
\Delta U = W_{on}.
$$

During compression, work is done *on* the gas, so $W_{on} > 0 \therefore \Delta U > 0$. This also corresponds to an increase in the gas's temperature. We recall from Eq. (4.38) that the work done *by* the gas during a quasi-static adiabatic process is:

$$
W_{by} = \frac{p_2V_2 - p_1V_1}{\gamma - 1}.
$$

## Question 3 - Increasing the temperature of a gas

An ideal gas is taken from a lower temperature $T_c$ to a higher temperature $T_h$. Since the ideal gas law is

$$
pV = nRT,
$$

an increase in temperature must be accompanied by a change in pressure, volume, or both.

#### Case 1: Constant Pressure (isobaric process)

If the pressure is held constant, $p = \text{constant}$, then the ideal gas law dictates

$$
V \propto T.
$$

As the temperature increases from $T_c$ to $T_h$, the volume increases. On a pV-diagram:
- The process is represented by a horizontal line.
- The system moves from left-to-right from a smaller volume to a larger volume.

#### Case 2: Constant Volume (isochoric process)

If the volume is held constant, then the ideal gas law dictates

$$
p \propto T.
$$

On a pV-diagram:
- The process is represented by a vertical line (constant $V$).
- The system moves **upwards** from a lower pressure to a higher pressure.

## Question 4 - Isothermal Processes

The process is **isothermal**, so the temperature remains constant at $T = 300\ \text{K}$.

For an ideal gas undergoing a reversible isothermal process, the work done *by* the gas is

$$
W_{by} = nRT \ \ln\left(\frac{V_2}{V_1}\right).
$$

Here,
- $n = 1\ \text{mol}$
- $R = 8.31\ \text{J mol}^{-1}\text{K}^{-1}$
- $T = 300\ \text{K}$
- $V_2 = 2V_1$

Substituting,
$$
W_{by} = (1)(8.31)(300)\ln(2).
$$

Evaluating,
$$
W_{by} \approx 1.73 \times 10^3\ \text{J}.
$$

Since the gas is expanding, the work done *by* the gas is positive. The work done *on* the system is therefore
$$
W_{on} = -W_{by}.
$$

### Final answer
$$
\boxed{
W_{on} \approx -1.7 \times 10^3\ \text{J}
}
$$

## Question 5 - Isothermal Expansion

#### Why doesn't the internal energy change?

The particles that constitute an ideal gas are assumed to be far apart from one another, and hence non-interacting. This results in a system whose energy is described solely by kinetic energy. which is directly related to temperature. If the temperature goes unchanged in a thermodynamic process, this indicates no adjustments to the gas particles' kinetic energy.

#### Work done by the gas during isothermal expansion.

Because the expansion of the ideal gas is isothermal, we can use $p = \frac{nRT_0}{V}$ and compute:

$$
W_{by} = \int_{V_1}^{V_2} p dV = \int_{V_1}^{V_2} \frac{nRT_0}{V} dV = nRT_0 \ \ln \left( \frac{V_2}{V_1} \right)
$$

For 1 mole ($n=1$),

$$
W_{by} = RT_0 \ \ln \left( \frac{V_2}{V_1} \right).
$$

#### Heat flow into the gas

Heat flow into the gas (i.e. $Q_{added}$) is simply equivalent to the work done *by* the gas. Using the first law we can easily solve for $Q_{added}$:

$$
\Delta U = 0 = Q_{added} - W_{by} \therefore Q_{added} = W_{by} = RT_0 \ \ln \left( \frac{V_2}{V_1} \right).
$$

## Question 6 - Heat Capacities and the Gas Constant

#### Step 1: Definitions

The molar heat capacities are defined as:

$$
C_V = \left(\frac{1}{n} \frac{\partial Q}{\partial T} \right)_V, \qquad C_p =  \left(\frac{1}{n} \frac{\partial Q}{\partial T} \right)_p.
$$

For an ideal gas, the adiabatic index is defined as

$$
\gamma = \frac{C_p}{C_V}.
$$

The gas constant $R$ is equal to the difference between the heat capacities at constant pressure and volume:

$$
C_p - C_V = R
$$

#### Step 2: Deriving $\tfrac{R}{C_V} = \gamma - 1$

Using $C_p = \gamma C_V$, substitute into the equation for $R$:

$$
\gamma C_V - C_V = R.
$$

Factorise out $C_V$:

$$
C_V(1 - \gamma) = R.
$$

Then finally divide both sides by $C_V$ to yield the desired expression:

$$
\boxed{\frac{R}{C_V} = \gamma - 1}
$$

#### Step 3: Deriving $\tfrac{R}{C_p} = \tfrac{\gamma - 1}{\gamma}$

Starting from $C_V (\gamma - 1) = R$, divide both sides of the equation by $C_p$:

$$
\frac{R}{C_p} = \frac{C_V(\gamma - 1)}{C_p}.
$$

Noting that $\gamma = \tfrac{C_p}{C_V}$, we also have $\tfrac{1}{\gamma} = \tfrac{C_V}{C_p}$, hence we can rewrite the above expression as:

$$
\boxed{
\frac{R}{C_p} = \frac{\gamma - 1}{\gamma}.
}
$$

## Question 7 - Cyclic Processes

#### (a) Why is the work negative?

On a $pV$-diagram, the net work done *by* the system over a complete cycle is given by the area enclosed by the loop:

$$
W_{by} = \oint p\,dV.
$$

Whether this work is positive or negative depends on the direction in which the cycle is followed.

Work is done *by* the system when it **expands**, because the gas *pushes* the piston outward. Work is done *on* the system when it is **compressed**, because the surroundings push the piston inward. The amount of work involved in either case depends on the pressure at which the expansion or compression occurs. **Pushing or being pushed at higher pressure involves larger forces.**

- For a **clockwise** cycle, the system expands at high pressures, then is compressed while the pressure is comparatively low. This means that the surroundings do less work on the gas during compression than the gas does on the surroundings during expansion. As a result, the net work done by the system over the cycle is positive.

- For a **counterclockwise** cycle, the system is compressed at high pressures, then expands under a lower pressure. Hence, the surroundings do more work on the gas during compression than the gas does during expansion and the net work done over the cycle is negative.

In this case, the path is counterclockwise, meaning that more work is done *on* the system than by it. This explains why the total work is negative.


#### (b) Change in internal energy and heat added

Because the process is **cyclic**, the system returns to its initial state at the end of the cycle. Internal energy is a **state function**, so it depends only on the state of the system and not on the path taken. As a result, the change in internal energy over a complete cycle is zero.

$$
\Delta U = 0.
$$

Using the first law of thermodynamics in the form

$$
\Delta U = Q + W_{on} = Q - W_{by},
$$

and noting that the given work is $W = -500\ \text{J}$ (interpreted as work done *by* the system), we have

$$
W_{by} = -500\ \text{J},
\qquad
Q_{added} = W_{by} = - 500 \ \text{J}.
$$

Therefore, 500 J of heat is removed during the cycle.

## Question 8 — Isobaric Boiling (UPMP – Example 19.5)

Given

- Initial volume: $V_1 = 1\ \text{cm}^3$
- Final volume: $V_2 = 1671\ \text{cm}^3$
- Constant pressure: $p = 1.013 \times 10^5\ \text{Pa}$
- Mass of water: $m = 1.0 \times 10^{-3}\ \text{kg}$
- Latent heat of vaporisation: $L_v = 2.256 \times 10^6\ \text{J kg}^{-1}$

First, convert the volumes to SI units. Since $1\ \text{cm}^3 = 10^{-6}\ \text{m}^3$,
$$
V_1 = 1 \times 10^{-6}\ \text{m}^3,
\qquad
V_2 = 1671 \times 10^{-6}\ \text{m}^3.
$$

The change in volume is therefore
$$
\Delta V = V_2 - V_1
= (1671 - 1)\times 10^{-6}
= 1.670 \times 10^{-3}\ \text{m}^3.
$$

#### (a) Work done *by* the water

At constant pressure, the work done by the system is simply the overall difference in volume, multiplied by the constant pressure:

$$
W_{by} = p\,\Delta V.
$$

Substituting the relevant values,
$$
W_{by} = (1.013 \times 10^5)(1.670 \times 10^{-3})
\approx 1.69 \times 10^2 \ \text{J}.
$$

$$
\boxed{
W_{by} \approx 169\ \text{J}
}
$$

### (b) Increase in internal energy

Notice the units of Latent Heat of vaporisation is joules per kilogram. To get the heat added, we can multiply this quantity by the mass of the water, which we know to be 1 gram. The heat added to the water during vaporisation is

$$
Q = m L_v
= (1.0 \times 10^{-3})(2.256 \times 10^6)
= 2.256 \times 10^3\ \text{J}.
$$

Using the first law of thermodynamics in the form

$$
\Delta U = Q + W_{on},
$$

and noting that $W_{on} = -W_{by}$, we obtain

$$
\Delta U = Q - W_{by}.
$$

Substituting the values we've calculated,

$$
\Delta U = 2256 - 169
\approx 2.09 \times 10^3\ \text{J}.
$$

$$
\boxed{
\Delta U \approx 2.09 \times 10^3\ \text{J}
}
$$

## Question 9 — Path Independence (TP - 11.3)

Consider the differential

$$
dz = 2xy \, dx + (x^2 + 2y)\, dy.
$$

We write this in the form

$$
dz = M(x,y)\,dx + N(x,y)\,dy,
$$

where

$$
M(x,y) = 2xy, \qquad N(x,y) = x^2 + 2y.
$$

#### (a) Path: $(x_1, y_1) \rightarrow (x_2, y_1) \rightarrow (x_2, y_2)$:

##### Segment 1: $(x_1, y_1) \rightarrow (x_2, y_1)$:

Along this segment, $y = y_1$ is constant, so $dy = 0$. The differential becomes

$$
dz = 2x y_1 \, dx.
$$

The contribution to the integral is

$$
\int_{x_1}^{x_2} 2x y_1 \, dx
= 2y_1 \int_{x_1}^{x_2} x \, dx
= 2y_1 \left[\frac{x^2}{2}\right]_{x_1}^{x_2}
= y_1 (x_2^2 - x_1^2).
$$

##### Segment 2: $(x_2, y_1) \rightarrow (x_2, y_2)$:

Along this segment, $x = x_2$ is constant, so $dx = 0$. The differential becomes

$$
dz = (x_2^2 + 2y)\, dy.
$$

The contribution to the integral is

$$
\int_{y_1}^{y_2} (x_2^2 + 2y)\, dy
= \left[x_2^2 y + y^2\right]_{y_1}^{y_2}
= x_2^2 (y_2 - y_1) + (y_2^2 - y_1^2).
$$

##### Total for path (a):

Adding both segments,

$$
\int_{\text{(a)}} dz
= y_1 (x_2^2 - x_1^2)
+ x_2^2 (y_2 - y_1)
+ (y_2^2 - y_1^2).
$$

This simplifies to

$$
\int_{\text{(a)}} dz
= x_2^2 y_2 - x_1^2 y_1 + (y_2^2 - y_1^2).
$$

#### (b) Path: $(x_1, y_1) \rightarrow (x_1, y_2) \rightarrow (x_2, y_2)$:

##### Segment 1: $(x_1, y_1) \rightarrow (x_1, y_2)$:

Along this segment, $x = x_1$ is constant, so $dx = 0$. The differential becomes

$$
dz = (x_1^2 + 2y)\, dy.
$$

The contribution to the integral is

$$
\int_{y_1}^{y_2} (x_1^2 + 2y)\, dy
= x_1^2 (y_2 - y_1) + (y_2^2 - y_1^2).
$$

##### Segment 2: $(x_1, y_2) \rightarrow (x_2, y_2)$:

Along this segment, $y = y_2$ is constant, so $dy = 0$. The differential becomes

$$
dz = 2x y_2 \, dx.
$$

The contribution to the integral is

$$
\int_{x_1}^{x_2} 2x y_2 \, dx
= y_2 (x_2^2 - x_1^2).
$$

##### Total for path (b):

Adding both segments,

$$
\int_{\text{(b)}} dz
= x_1^2 (y_2 - y_1)
+ (y_2^2 - y_1^2)
+ y_2 (x_2^2 - x_1^2).
$$

This simplifies to

$$
\int_{\text{(b)}} dz
= x_2^2 y_2 - x_1^2 y_1 + (y_2^2 - y_1^2).
$$

#### Is $dz$ an exact differential?

For a differential of the form

$$
dz = M(x,y)\,dx + N(x,y)\,dy,
$$

the differential is exact if

$$
\frac{\partial M}{\partial y} = \frac{\partial N}{\partial x}.
$$

Here,

$$
\frac{\partial M}{\partial y} = \frac{\partial}{\partial y}(2xy) = 2x,
$$

and

$$
\frac{\partial N}{\partial x} = \frac{\partial}{\partial x}(x^2 + 2y) = 2x.
$$

Since these are equal, $dz$ is an exact differential. The value of the integral therefore depends only on the start and end points, not on the path taken.
