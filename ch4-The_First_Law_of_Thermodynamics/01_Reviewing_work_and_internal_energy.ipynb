This chapter introduces the **first law of thermodynamics**, the central bookkeeping principle that governs all thermal processes. At its core, the first law expresses energy conservation for thermodynamic systems, linking changes in a system's internal energy to energy transferred by **heat** and **work**. Throughout the chapter, we will sharpen our definitions of heat, work and internal energy, establishing clear sign conventions used in subsequent derivations. This notational framework will allow us to analyse physical processes consistently, regardless of the complexity of the path between states.

Building on this foundation, this chapter will explore how different **thermodynamic processes** - such as **cyclic**, **isothermal**, **isochoric**, **isobaric** and **adiabatic** transformations - emerge from the first law when specific constraints are imposed. We then examine **heat capacities**, showing how they quantify a system's response to heating and lead naturally to the gas constant $R$ and the adiabatic index $\gamma$. These ideas culminate in the derivation of the **adiabatic equation of state**, $pV^{\gamma} = \text{const}$, revealing how microscopic energy storage shapes macroscopic gas behaviour.

We will begin by reviewing the concepts of work and internal energy.

## 4.1 Review: Work

The conservation of energy is fundamental to thermodynamics because it governs how energy is transferred and transformed within a system. It provides the basis for the first law of thermodynamics, which relates changes in internal energy to heat added to a system and work done by the system. By applying energy conservation, we can distinguish between energy stored internally and energy exchanged as work or heat, enabling a consistent understanding of thermodynamic processes and their limits.

### Work:

Work is done by exerting a force on an object, whilst the object is moved from one position to another (or has its shape warped). There is a displacement of the object (see figure below):

[FIGURE 6.2 HERE]

Work measures the energy transferred by a force acting through a process of displacement. When an object moves a distance $s$ in a straight line under a constant force $F$ which acts in the same direction as the displacement, the work done is calculated to be simply the product of $F$ and $s$:

$$
W = F s
\tag{4.1}
$$

$$
\text{[Joules, J]} = \text{[Newtons, N]} \cdot \text{[Displacement, m]}
$$

From the equation above we can see that work can be increased by amplifying the force or by applying the same force over a greater distance.

### Sign of work:

The sign of the work depends on the relative angle of the force to the object that is being displaced. The figure below helps to illustrate how this works:

[FIGURE 6.3 HERE]

With vector representations for the quantities involved, $\vec{F}$ & $\vec{s}$ we can produce two other equivalent statements for work:

$$
W = |\vec{F}||\vec{s}| \ \cos(\phi) = Fs \ \cos(\phi),
\tag{4.2}
$$

where $\phi$ is the angle between $\vec{F}$ and $\vec{s}$. If $\phi=0$, so that $\vec{F}$ and $\vec{s}$ act in the same direction, then $\cos(\phi)=1$ and we regain Eq. (4.1).

Eq. (4.2) has the form of a *scalar product* of two vectors, meaning it can be equivalently represented as a dot product of two vectors:

$$
W = \vec{F} \cdot \vec{s}.
\tag{4.3}
$$

Work can take any real value, from $-\infty$ to $+\infty$, but it is a *scalar* quantity: it has magnitude but no intrinsic direction. This is true despite work being defined in terms of vector quantities. A 5 N force acting eastward on an object that then moves 6 m east does the same amount of work as a 5 N force acting northward on an object which then moves 6 m north.

The sign of the work encodes the direction of energy transfer between the system and its surroundings. Once a sign convention is fixed, positive and negative values of work correspond to different physical situations. In this convention, positive work represents energy transferred to the system by the surroundings, while negative work represents energy transferred from the system to the surroundings. The mathematical form of the first law remains unchanged; only the interpretation of the sign differs.

### The work-energy theorem

The total work done on an object by external forces is linked to its displacement, but its physical importance lies in how it changes the object's speed. Work is positive when the net force has a component in the direction of motion, negative when it opposes the motion, and zero when the net force lies perpendicular to the displacement (or when no force is applied, of course).

The effect of work on motion:
- Positive work $\rightarrow$ Increases the object's speed.
- Negative work $\rightarrow$ Decreases the object's speed.
- Zero work $\rightarrow$ The object's speed is unchanged.

Thus, work provides a direct quantitiative measure of how forces modify motion. The kinetic energy of a particle is defined as:

$$
K = \tfrac{1}{2} m v^{2},
\tag{4.4}
$$

$$
\text{[Joules, J]} = \text{[Mass, kg]} \cdot \text{[Velocity, m/s]}^2
$$

where $m$ is the mass of said particle, and $v$ is its' velocity. Like work, kinetic energy is a scalar quantity, independent of the direction of motion; although in this case it is always non negative ($K \geq 0$), only becoming zero when the particle is at rest.

For motion along a straight line under a constant net force, combining Newton's second law with kinematics shows that the work done by the net force over a displacement equals the difference between the final and initial kinetic energies:

$$
W_{tot} = K_{final} - K_{initial} = \Delta K.
\tag{4.5}
$$

This is obtained via the following derivation:

$$
v_{f}^2 = v_{i}^2 + 2as,
$$

where $v_f$, $v_i$ denote the final and intial particle velocity, whilst *a* denotes the acceleration corresponding to the applied force ($F=ma$). We can rearrange the equation above to get an expression for the displacement:

$$
s = \frac{v_{f}^2 - v_{i}^2}{2a}.
$$

The work of the net force is the product of its magnitute ($F=ma$) and the particle's displacement (assuming the vectors $\vec{F}$ and $\vec{s}$ are parallel):

$$
\begin{align*}
W = Fs &= ma \frac{v_f^2 - v_i^2}{2a} = \tfrac{1}{2} m v_f^2 - \tfrac{1}{2} m v_i^2 \\
  &= K_f - K_i = \Delta K
\end{align*}
$$

This theorem states that the total work done by all forces acting on a particle simply equals its overall change in kinetic energy. If the net work is positive, $\Delta K$ is also positive and the particle speeds up. The reverse is true if work is negative, and of course if work is zero then the particle's speed is unchanged even though forces may be acting. Although derived for a straight-line motion with constant forces, the work-energy theorem is valid for more general motions and force laws, provided the analysis is carried out in an intertial reference frame.

Note that if an object is displaced from rest (e.g. a billiard ball), then the work-energy theorem is simply:

$$
W = K_2 - 0 = \tfrac{1}{2}mv^2.
$$

### Work and energy with varying forces:

Above, work was defiend for constant forces. In many physical situatuions however, the force varies in magnitude and/or direction along the path of motion. The work-energy theorem remains valid even in these cases.

Consider motion along the x-axis, where the x-component of the force, $F_x(x)$, varies with position. We can divide the displacement from $x_1$ to $x_2$ into small segments $\Delta x$, where we can linearly approximate the work done over each segment:

$$
\Delta W = Fx \Delta x;
$$

this would allow us to approximate the total work by summing over each of these segments...

$$
W = Fx_1 \Delta x_1 + Fx_2 \Delta x_2 + ...
$$

The exact expression is given in integral form, in the limit of $\Delta x \rightarrow 0$:

$$
W = \int_{x_1}^{x_2} F_x(x) dx.
\tag{4.6}
$$

Graphically, this work equals the area under the $F_x(x)$ curve between $x_1$ and $x_2$. For a constant force $F_x$, this reduces to $W = F_xx(x_2 - x_1)$.

Therefore, the work-energy theorem indeed holds for scenarios where the driving force varies with position.

We can also derive this outcome by considering a calculus approach. The acceleration of a particle can be described using the *chain rule*:

$$
a_x = \frac{dv_x}{dt} = \frac{dv_x}{dx}\frac{dx}{dt} = \frac{dv_x}{dx}v_x
$$

Using Newton's second law to write $F_x = ma_x$, the total work is given by:

$$
W_{tot} = \int_{x_1}^{x_2} F_x dx = \int_{x_1}^{x_2} ma_x dx = \int_{x_1}^{x_2} mv_x \frac{dv_x}{dx}dx
$$

The $dx$ in the numerator and denominator cancel out, making leaving the integration with respect to velocity:

$$
W_{tot} = \int_{v_1}^{v_2} mv_x dv_x = \tfrac{1}{2}mv_2^2 - \tfrac{1}{2}mv_1^2
$$

which yields the same conclusion as Eq. (4.5). Therefore, the work-energy theorem is valid for straight-line motion even when the net force varies with position, and does not rely on the assumption of constant force.

We can also extend the work-energy theorem for motion along a curve, in which a force varies in direction as well as magnitude, and thus the displacement lies along a curved path. Consider the scenario depicted in the figure below:

[FIGURE 6.23]

For a particle moving from point $P_1$ to $P_2$ along a curve:
- Divide the path into infinitesmal vector displacements $dl$, each tangential to the path.
- Let $F$ be the force at a point along the path.
- Only the component of the force parallel to the displacement does work.

The differential work is:

$$
dW = \vec{F} \cdot d\vec{l} = F\cos(\phi) dl = F_{\parallel}dl
$$

Phi is the angle between $F$ and $dl$, and $F_{\parallel} = Fcos(\phi)$ is the force component along the path. The total work done as the particle moves from P1 to P2 is therefore:

$$
W = \int_{P_1}^{P_2} = \vec{F} \cdot d\vec{l} = \int_{P_1}^{P_2} F\cos(\phi) \ dl = \int_{P_1}^{P_2} F_{\parallel} dl.
\tag{4.7}
$$

This is a line integral, and it accounts for both the changing direction of motion and the changing force along the path. The work-energy theorem holds generally - by summing up the infintesimal work $dW = Fdl$ over all segments of a curved path, the total work equals the total change in kinetic energy $W_{tot} = \Delta K = K_2 - K_1$, regardless of how the force varies.

Only the component of the force *parallel to the path* does work and changes the particle's speed, while the perpendicular component changes only in the direction of motion.
