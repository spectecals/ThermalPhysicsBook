## Questions

1. The theory of equipartition relates the internal energy of a gas to its temperature.

   1. What assumptions are made in the theory of equipartition? (2)
   2. Which of these assumptions must be modified for real gases, and why? (2)
   3. Explain the implications of the theory of equipartition for monatomic versus diatomic molecules. (2)
   4. State the internal energy of:
      - one mole of a monatomic gas, and  
      - one mole of a diatomic gas,  
      assuming a temperature of 300 K. (2)
   5. What are the implications of the theory of equipartition for gases at low and high temperatures? (2)

2. The pressure $p$ in an ideal gas can be shown to be
   $$
   p = \frac{2}{3} N_V \overline{E_k},
   $$
   where $N_V$ is the number of molecules per unit volume and $\overline{E_k}$ is the average kinetic energy of a molecule. By relating this expression to the pressure obtained from the ideal gas law, what can be concluded about the energy per degree of freedom in an ideal gas? (5)

3. Consider a gas that is raised from temperature $T_C$ to $T_H$. This process is conducted either at constant pressure or at constant volume. By considering these two processes, show that the ratio of the molar specific heat capacities at constant pressure $c_p$ and constant volume $c_V$ of an ideal gas is equal to $5/3$. (5)
