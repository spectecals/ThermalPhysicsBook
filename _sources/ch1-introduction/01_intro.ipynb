## What is a mole?

In thermal physics, we deal with enormous numbers of atoms and molecules. These large numbers make it impossible to track individual particles, so we describe their behaviour statistically, in terms of averages. Even numbers we think of as huge, like the Earth’s population (~$10^9$) or the U.S. national debt (~$10^{13}$), are tiny compared with the numbers of atoms in ordinary matter (around $10^{23}-10^{25}$ per sample). For instance, 1 kg of nitrogen gas contains about $2*10^{25}$ molecules. A fast computer counting one molecule per clock cycle would take hundreds of millions of years just to count them! To handle such vast quantities, we use the concept of the mole, a convenient counting unit similar to a "dozen", but much larger.

```{admonition} Definition:
:class: definition
A mole is the quantity of matter that contains as many objects (atoms, molecules, ions) as there are atoms in *exactly* 12 grams of $^{12}$C.
```

This number, called **Avogadro’s number**, is $N_A=6.022×10^{23}$. Thus, one mole of carbon, water, or sodium chloride each contains $6.022×10^{23}$ of their respective particles.

## Useful constants

Other than Avogadro's number, there are other definitions useful to understand:

The **Boltzmann *constant***, $k_B$
- $k_B=1.38*10^{-23}$ J K<sup>-1</sup>
- This constant relates the average energy $J$ of a particle in a system to its temperature $K$. It serves as a conversion factor between temperature and energy.

The **universal gas *constant***, $R$
- $R = 8.31$ J K<sup>-1</sup> mol<sup>-1</sup>,
- Like the Boltzmann constant, this is a proportionality constant relating the energy $J$ and temperature $K$ and the amount of substance used $\text{mol}$. In other words, it's the molar equivalent to the Boltzmann constant. This is the constant of proportionality used in the macroscopic form of the ideal gas law (CTP Section 1.4).
- i.e., $R = N_A*k_B = k_B * 6.022×10^{23} $ mol<sup>-1</sup>.

## Systems

We refer to our objects or regions of interest as a **system**. Everything outside if this system is their **surroundings**, where the **boundary** is the real or imaginary surface that separates the two. The interaction of the system across the boundary with their surroundings define the type of system we're working with.

- **Open systems**: An open system can exchange both matter and energy with its surroundings.
- **Closed systems**: A closed system cannot exchange matter with its surroundings, but energy can still cross the bounday.
- **Isolated systems**: An isolated system cannot exchange matter nor energy with its surroundings.

Thermodynamics is a field that describes its systems while in thermal equilibrium with its surroundings (no heat flow), allowing us to describe changes as transitions between one equilibrium state to another. Each equilibrium state is fully specified by **state variables**, which have definite values only while in equilibrium.

```{admonition} Definition:
:class: definition
A system is in <b>thermal equilibrium</b> when its temperature is the same throughout and does not change with time.
```

If the system is in contact with its surroundings, thermal equilibrium means there is no net heat flow across the boundary. Under these conditions, macroscopic properties remain stable which is why they can be fully described by state variables.


## State variables

A quantity with a unique value for a particular state of the system is a *state variable*, i.e.:

- $V$, volume
- $U$, internal energy
- $p$, pressure, the force per unit area: $p = F/A$
    - There is a zoo of units used to express pressure, which can be confusing for beginners since different fields favour different conventions. Below are some of the commonly used units that you may encounter:
        - SI units:
            - pascals (Pa); 1 Pa = 1 N m<sup>-2</sup>. This is the official unit for pressure, and the one we will use in this course
        - Non-SI units: 
            - bars; 1 bar = 100,000 Pa
            - atmospheres (atm); 1 atm = 101,325 Pa. Originally defined as the average atmospheric pressure at sea level
            - pounds per square inch (psi); 1 psi = 6,895 Pa. Imperial unit, commonly used in US industries, i.e., tire pressure
            - torrs (Torr); 1 Torr = 133.32 Pa = 1 atm / 760. *Slightly* (0.000 015%) smaller than the millimetre of mercury (mmHg), widely used in medicine

If you know the state, these values are fixed, no matter how you got there. Alternatively, some variables implicitly record the history of how a system has changed, these are definitively *not* state variables:

- $Q$, heat (involves energy transfer to/from other systems)
- $W$, work (i.e., pushing something, powering a motor).

## Intensive vs extensive variables

### Intensive variables

These are variables that do not scale with the size of your system. Some examples are:

- $p$, pressure
- $T$, temperature
- $\rho$, density
- $v$, specific volume, the volume per unit mass: $v = \frac{V}{m}$

### Extensive variables

These are variables that **do** scale with the size of your system. Consider:

- $V$, volume
- $m$, mass
- $n$, number of moles
- $U$, internal energy

If you drank half a cup of warm coffee, you'll obviously find that the number of atoms, volume, internal energy and the heat it radiates all halve.

## Changing thermodynamic states

A process that changes a system's state is described by a **path**. The path describes *how* a system changes from one equilibrium state to another. If a process can or cannot be undone, we consider these processes and paths either "reversible" or "irreversible". 

A given pair of equilibrium states can be connected by many distinct paths. For example, a system may be taken from state A to state B by holding temperature constant while varying pressure, or by holding pressure constant while varying temperature, or by varying these variables sequentially or in some combination. The path therefore describes the conditions maintained during the process, not just the initial and final states.

```{figure} ../src/figures/1.5_paths.png
:name: fig:paths
Both processes start and end in the same state, though their paths differ significantly. Path 1 increases $p$ while holding $T$ constant, then increases $T$ while $p$ is constant. Path 2 keeps $p / T$, or equivalently $V$, (roughly) constant.
```

A **reversible** path has the system remain in equilibrium at all times. This requires the process to proceed sufficiently slowly that the system can continually adjust to changing conditions. In the ideal limit, the path consists of an infinite sequence of equilibrium states, so that the state variables are well defined at every stage of the process.

An **irreversible** path involves changes that drive the system away from equilibrium such that they can no longer be described by state variables. Consider the rapid compression of a gas with a piston, vigorous stirring, or expansion into a vacuum. During such processes, different parts of the system may not be in equilibrium with one another. As a result, state variables cannot describe the system throughout the process, and the path itself cannot be specified in terms of these variables.

<!-- 
```{figure} ../src/figures/1.5_reversible.png
:name: fig:reversible
A reversible process must be perfectly lossless. Real processes aren't *truly* reversible but some get close.
``` -->

Paths are often classified by which physical quantity is constrained during the process:

- **isothermal**: temperature remains constant
- **isobaric**: pressure remains constant
- **adiabatic**: no heat is exchanged with the surroundings


<!-- ## The thermodynamic limit

Thermodynamics is a phenomenological theory of macroscopic variables and therefore only strictly holds true for large systems. In fact, we must begin by assuming that the number of particles, $N$, tends to infinity, as does the volume $V$, in such a way that the density $\rho$ of the gas remains constant and uniform,

$$N \rightarrow \inf, \qquad V \rightarrow \inf, \qquad \rho = \frac{N}{V}. $$ -->



(sec:idealgas)=
## Ideal gases

This section introduces a very important concept called the *ideal gas law*, based around the understanding that the pressure of a gas depends on its temperature and volume.

- **Boyle's law** (who boiled a viper in a vacuum): Experiments show that, at constant temperature, the product of $p V$ remains constant,

$$ p \propto \frac{1}{V}. $$

- **Charles' law** (also an avid balloonist): When the pressure is held constant, the volume of a gas increases proportionally with temperature,

$$ V \propto T. $$

- **Gay-Lussac's law** (who discovered the composition of water): At constant volume, the pressure is proportional to temperature,

$$ p \propto T. $$

Combining these empirical results, alongside $\rho = \frac{N}{V}$, we find that $p V \propto N T$. Their proportionality is the aforementioned Boltzmann constant, which relates temperature to energy, and forms the **microscopic ideal gas equation**,

$$ p V = N k_B T \qquad \text{(microscopic)}. $$

Alternatively, we can define it macroscopically in terms of the numbers of moles, $n$, through $N=n * N_A$ and the universal gas constant, $R = N_A*k_B$, forming the **macroscopic ideal gas equation**,

$$ p V = n R T \qquad \text{(macroscopic)}. $$

This *ideal gas* model assumes that molecules:
1. move randomly,
2. collide elastically,
3. do not have intermolecular interactions (such as attracting or repelling each other),
4. and are point-like with negligible volume.

Under these assumptions, the gas pressure arises from molecules striking the container walls. Despite its simplicity, the model accurates describes many real gases under conditions of low pressure and high temperature. At higher densities or lower temperatures, deviations occur because real molecules have finite size and experience intermolecular forces.

```{figure} ../src/figures/1.6_kinetic.png
:name: fig:kinetic_billiards
Ideal gas as billiard balls in a box. Gas particles are modeled as identical point-like balls moving randomly and colliding elastically with each other and with the walls of the container, illustrating how microscopic motion gives rise to state variables such as pressure and temperature.
```

## Combinatorial problems

Thermal physics studies how large collections of particles behave *statistically*. Since it is impossible to track every particle individually, we instead count how many distinct configurations a system can have that correspond to a given macroscopic state. This kind of counting is a combinatorial problem specifically relevant for indistinguishable particles.

### The combination formula

When counting configurations, we need to know how many ways we can arrange a certain number of items from a larger set **without regard to order**. The number of such combinations $C$ is given by:

$$ ^nC_k = \frac{n!}{k!(n-k)!} $$

Here $n$ is the total number of available positions, $k$ the number of elements we choose to occupy them and $n!$ refers to "$n$ factorial", meaning $n * (n - 1) * (n - 2) *\ ...\ * 1$. This formula tells us how many distinct configurations are possible when order does not matter.

```{admonition} Example: (Identical) eggs in a box
:class: example

Imagine we have a carton able to hold 12 **indistinguishable** eggs. How many distinct arrangements of this system are possible if the system has:

- (a) 12 eggs?
- (b) 4 eggs?

We can approach these problems by using $ ^nC_k = \frac{n!}{k!(n-k)!}$ where $n$ here is the number of spaces in our carton and $k$ is the number of eggs. 

## (a)

As you can imagine, there's only one way to store 12 eggs in a carton with 12 spaces:

$$ \frac{n!}{k!(n-k)!} = \frac{12!}{12!(12-12)!} = \frac{12!}{12!*0!} = \frac{12!}{12!*1} = 1. $$

A critical point of note here, $0! = 1$. This may not make sense at first but you can make sense of it by considering how many ways you can arrange 0 items. There's one way&mdash;by not arranging them at all!

## (b)

Using the same approach as before with $ ^nC_k $:

$$ \frac{n!}{k!(n-k)!} = \frac{12!}{4!(12-4)!} = \frac{12!}{4!*8!} = \frac{12 * 11 * 10 * 9}{4 * 3 * 2} = 495. $$
```

Note that for case (b), the intuitive approach "there are 12 spaces for the first egg, then 11 for the second," so on and so forth, would lead you to the **incorrect** answer for the number of combinations,

$$(\text{from intuition, or}\ ^nP_k) \qquad 12 * 11 * 10 * 9 = \frac{12!}{8!}  \neq \frac{12!}{4!*8!}\qquad(\text{from}\ ^nC_k) .$$

The equation from "intuition" is the **permutation formula** $ ^nP_k = \frac{n!}{(n-k)!} $ which is used in situations where the ordering matters (i.e., a pack of cards). This discrepancy between solutions arises because **the eggs are indistinguishable**, meaning that certain arrangements are repeats of prior combinations. This is **always the case in thermodynamics** as it's not possible to identify and keep track of each particle in a gas.

(If you *really* wanted to use intuition, you would need to implement an additional condition: "how many ways can I arrange each of my permutations?" For any given arrangement of the 4 eggs, you could rearrange them amongst themselves like so: "there are 4 spaces for the first egg, 3 for the second, ..." resulting in $4!=12$ repetitions of each distinct arrangement. By dividing the permutations by the repetitions, you rederive the combinatorial solution!)
