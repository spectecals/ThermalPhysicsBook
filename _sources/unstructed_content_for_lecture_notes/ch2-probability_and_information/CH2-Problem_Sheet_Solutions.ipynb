## Solution to Question 1

Option 1 gives you a certain £10.

Option 2 gives you £36 only if the die shows 5 or 6. Since the die is fair:

$$
P(5 \text{ or } 6) = \frac{1}{6} + \frac{1}{6}
= \frac{2}{6}
= \frac{1}{3}.
$$

The expected winnings are:

$$
\langle \text{£} \rangle = \frac{1}{3} \times 36 = 12.
$$

Since £12 $>$ £10, **Option 2 is the better economic choice**.

## Solution to Question 2

We want the probability that you actually have the disease given that the test is positive:

$$
P(D \mid +).
$$

Using Bayes' theorem:

$$
P(D \mid +)
=
\frac{P(+ \mid D)\,P(D)}
{P(+ \mid D)\,P(D) + P(+ \mid \bar{D})\,P(\bar{D}) }.
$$

Given:

- $P(+ \mid D) = 0.98$
- $P(+ \mid \bar{D}) = 0.02$
- $P(D) = \frac{1}{10000} = 0.0001$
- $P(\bar{D}) = 0.9999$

Substitute:

$$
P(D \mid +)
=
\frac{0.98 \times 0.0001}
{0.98 \times 0.0001 + 0.02 \times 0.9999}.
$$

$$
P(D \mid +)
\approx
\frac{0.000098}
{0.000098 + 0.019998}
\approx 0.00488.
$$

So the probability you actually have the disease is:

$$
P(D \mid +) \approx 0.488\%.
$$

## Solution to Question 3

Let the hypotheses be:

1. $H_{1}$: both balls red
2. $H_{2}$: one red, one green
3. $H_{3}$: both balls green

Initially:

$$
P(H_{1}) = P(H_{2}) = P(H_{3}) = \frac{1}{3}.
$$

You draw a ball out of the bag and see that it is green (we denote this event $G$, and red $R$).
Based on this outcome, we can compute the probabilities of obtaining $G$ given the hypotheses, and thus the new probabilities for the three hypotheses.

### Updating $H_{1}$:

Since we obtained a green ball, hypothesis $H_1$ cannot be true ($P(H_{1} \mid G) = 0$). It is fair to say that the inverse is true. If hypothesis 1 is correct, and the balls are all red, then the probability of the ball being green would be zero ($P(G \mid H_{1}) = 0$).

### Updating $H_{2}$

Hypothesis 2 conjectures that there is a ball of each colour in the bag. Since we have observed a green ball. In this framework, the probability of obtaining a ball from this bag, $P(G \mid H_2)$ would be 0.5. We can thus compute the posterior via Bayes' Theorem:

$$
P(H_2 \mid G) = \frac{P(G \mid H_2)P(H_2)}{P(G)} = \frac{\tfrac{1}{2} \cdot \tfrac{1}{3}}{\tfrac{1}{2}} = \frac{1}{3}
$$

### Updating $H_{3}$

Finally, we can update our hypothesis that both balls in the bag are green ($H_3$) in an identical fashion to the previous hypothesis, noting only this time that $P(G \mid H_3) = 1$:

$$
P(H_3 \mid G) = \frac{P(G \mid H_3)P(H_3)}{P(G)} = \frac{1 \cdot \tfrac{1}{3}}{\tfrac{1}{2}} = \frac{2}{3}
$$

## Solution to Question 4

Given:

$$
p_{h} = 0.3,
\qquad
n = 5.
$$

We want the probability of $k=2$ heads:

$$
P(k=2)
=
\frac{n!}{k!(n-k)!}
p_{h}^{k}
(1-p_{h})^{n-k}.
$$

Substitute:

$$
P(2)
=
\frac{5!}{2!\,3!}
(0.3)^{2}
(0.7)^{3}
\approx 0.3087.
$$

So the probability is **30.9%**.

```python
# Parameters
p = 0.3          # probability of heads
n = 5            # number of trials
k_vals = np.arange(0, n+1)

# Binomial probabilities
probs = np.array([comb(n, k) * p**k * (1-p)**(n-k) for k in k_vals]) * 100

# Plot
plt.figure(figsize=(6,4))
markerline, stemlines, baseline = plt.stem(k_vals, probs)

plt.setp(markerline, marker='o', markersize=6, color='black')
plt.setp(stemlines, linewidth=2, color='black')
plt.setp(baseline, linewidth=0)

plt.xlabel("k (number of heads)")
plt.ylabel("P(k) (%)")
plt.title("Binomial Distribution: n = 5, p = 0.3")
plt.ylim(0, max(probs)*1.2)

for k, p_k in zip(k_vals, probs):
    plt.text(k, p_k + 1.5, f"{p_k:.2f}", ha='center')

plt.tight_layout()
plt.show()
```

### Mean and standard deviation for 10,000 tosses

$$
\mu = np = 10000 \times 0.3 = 3000 \ \text{Heads},
$$

$$
\sigma = \sqrt{np(1-p)}
=
\sqrt{10000 \times 0.3 \times 0.7}
\approx 45.8 \ \text{Heads}.
$$

## Solution to Question 5

The Poisson distribution is the limiting case of the binomial when:

- $n \to \infty$
- $p \to 0$
- $np = \lambda$ stays constant.

Poisson Probability Mass Function (PMF):

$$
P(k) = \frac{\lambda^{k} e^{-\lambda}}{k!}.
$$

### Probability of avoiding a plane crash after 1000 flights

Probability of a crash per flight:

$$
p = 10^{-5}.
$$

Number of flights:

$$
n = 1000.
$$

Expected crashes:

$$
\lambda = np = 1000 \times 10^{-5} = 0.01.
$$

Probability of zero crashes:

$$
P(0) = \frac{\lambda^{0} e^{-\lambda}}{0!} = e^{-0.01}
\approx 0.99005.
$$

So the probability of surviving all 1000 flights is:

$$
\approx 99.0\%.
$$
