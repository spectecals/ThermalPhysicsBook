# Welcome to Thermal Physics!

Dr. Samer Kurdi & Dr. Jonathan Leach


## Course Syllabus

This course will further develop your understanding in thermal physics in the following concepts:

- **Laws of Thermodynamics**: Apply the zeroth, first, and second laws of thermodynamics to a range of physical systems.
- **Probability, Entropy, and Statistical Foundations**: Understand how probability and entropy underpin thermodynamics and statistical mechanics.
- **Thermal Properties of Matter**: Understand the properties and behaviour of matter, including phase changes, phase diagrams, heat capacities, and thermal expansion.
- **Kinetic Theory and Real Gases**: Understand the kinetic theory of gases and its application to real gas behaviour.
- **Heat Engines and Thermodynamic Cycles**: Analyze heat engine cycles and apply the laws of thermodynamics to their operation.

## Context

The laws of thermodynamics summarize the properties of energy and its transformation from one form to another, providing a unifying understanding of the role of energy in the universe. Although thermodynamics emerged in the 19th century from the study of steam engines and efficiency limits, its relevance extends far beyond classical engines. It governs the operation of heat pumps and refrigerators, defines the ultimate efficiency of energy conversion, and is central to state-of-the-art technologies such as millikelvin refrigerators used in quantum experiments. More broadly, thermodynamics underpins modern quantum technologies, where controlling energy flow, dissipation, and thermal noise is essential for achieving and preserving quantum behaviour.
