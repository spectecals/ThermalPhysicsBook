## Solutions

1. The world’s power consumption is currently around 20 TW, and growing!
(1 TW = 10^12 W.) Burning one ton of crude oil (which is nearly seven barrels)
produces about 42 GJ (1 GJ = 10^9 J). If the world’s total power needs were to
come entirely from burning oil (a large fraction currently does), how much oil
would we be burning per second?

$$
\frac{20 \times 10^{12}\ \text{J s}^{-1}}{42 \times 10^{9}\ \text{J tonne}^{-1}}
= 4.76 \times 10^{2}\ \text{tonnes s}^{-1}
$$


2. The molar heat capacity of gold is
25.4 J mol^{-1} K^{-1}. Its density is
19.3 \times 10^{3} kg m^{-3}.
Calculate the specific heat capacity of gold and the heat capacity per unit
volume. What is the heat capacity of 4 \times 10^{6} kg of gold?
(This is roughly the holdings of Fort Knox.)

1 mole of Au has mass 197 g = 0.197 kg.

Specific heat capacity:
$$
c = \frac{25.4\ \text{J mol}^{-1}\text{K}^{-1}}{0.197\ \text{kg mol}^{-1}}
= 1.29 \times 10^{2}\ \text{J kg}^{-1}\text{K}^{-1}
$$

Heat capacity per unit volume:
$$
C_V = \rho c
= (19.3 \times 10^{3})(1.29 \times 10^{2})
\approx 2.5 \times 10^{6}\ \text{J m}^{-3}\text{K}^{-1}
$$

Heat capacity of 4 \times 10^{6} kg of gold:
$$
C = mc
= (4 \times 10^{6})(1.29 \times 10^{2})
\approx 5.2 \times 10^{8}\ \text{J K}^{-1}
$$


3. Explain the differences and similarities between the definitions of the
kelvin and Celsius temperature scales. (5)

- Same interval size: a change of 1 K equals a change of 1 ^\circ C.
- Different zero points: kelvin is an absolute scale; Celsius is offset.
- Conversion:
  $$
  T(\text{K}) = T(^\circ\text{C}) + 273.15
  $$
- Physical meaning of zero:
  - 0 K is absolute zero.
  - 0 ^\circ C was historically defined by the freezing point of water at 1 atm
    (now defined by a fixed offset from kelvin).


4. The triple point of CO_2 is at 5.11 atm and 216.6 K. Describe the phase
changes that occur in a sample of CO_2 when:

- The temperature increases from 200 K to 273 K at 1 atm.
- The temperature increases from 200 K to 273 K at 100 atm.
- The pressure is reduced from 10 atm to 1 atm at 273 K.

Include sketches to show these phase changes relative to the phase boundaries. (5)

Triple point of CO_2:
$$
P_t = 5.11\ \text{atm}, \qquad T_t = 216.6\ \text{K}.
$$
Below the triple-point pressure, liquid CO_2 cannot exist.

(a) Heating from 200 K to 273 K at 1 atm

- 1 atm < 5.11 atm, so no liquid phase is possible.
- At 200 K, CO_2 is solid.
- On heating, the solid–gas (sublimation) boundary is crossed.
- At 273 K, CO_2 is a gas.

(b) Heating from 200 K to 273 K at 100 atm

- 100 atm > 5.11 atm, so a liquid phase is possible.
- At 200 K, CO_2 is solid.
- Heating crosses the solid–liquid boundary (melting).
- At 273 K and 100 atm, CO_2 remains liquid.

(c) Reducing pressure from 10 atm to 1 atm at 273 K

- At 273 K, CO_2 is gaseous at both 10 atm and 1 atm.
- No phase boundary is crossed; the sample remains a gas.

The three paths on a P–T diagram are:
- (a) a horizontal line at 1 atm crossing the sublimation line,
- (b) a horizontal line at 100 atm crossing the melting line,
- (c) a vertical line at 273 K remaining in the gas region.


5. What is the minimum quantity of ice at -10 ^\circ C required to bring
500 g of liquid water at 60 ^\circ C to 0 ^\circ C? (5)

Use:
c_water = 4180 J kg^{-1} K^{-1}
c_ice   = 2100 J kg^{-1} K^{-1}
L_f     = 334000 J kg^{-1}

Heat released by cooling the water:
$$
Q_{\text{water}} = mc\Delta T
= (0.500)(4180)(60)
= 1.254 \times 10^{5}\ \text{J}.
$$

Heat absorbed per kg of ice:

Warm ice from -10 to 0 ^\circ C:
$$
q_1 = c_{\text{ice}}(10) = 2.10 \times 10^{4}\ \text{J kg}^{-1}
$$

Melt ice at 0 ^\circ C:
$$
q_2 = L_f = 3.34 \times 10^{5}\ \text{J kg}^{-1}
$$

Total per kg:
$$
q = q_1 + q_2 = 3.55 \times 10^{5}\ \text{J kg}^{-1}
$$

Equating heat exchanged:
$$
m_{\text{ice}} q = Q_{\text{water}}
$$

$$
m_{\text{ice}} = \frac{1.254 \times 10^{5}}{3.55 \times 10^{5}}
\approx 0.35\ \text{kg}.
$$

Minimum ice required: approximately 350 g.
