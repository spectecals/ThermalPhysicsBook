## Solutions

1. The world’s power consumption is currently around 20 TW, and growing! (1 TW = 10<sup>12</sup>W.) Burning one ton of crude oil (which is nearly seven barrels worth) produces about 42 GJ (1 GJ= 10<sup>9</sup>J). If the world’s total power needs were to come entirely from burning oil (a large fraction currently does), how much oil would we be burning per second?

$$ \frac{20 \cdot 10^{12} \text{W}}{42 \cdot 10^9 \text{J/ton crude}} \approx 450 \text{tons of crude oil s}^{-1} $$

2. The molar heat capacity of gold is 25.4 Jmol<sup>−1</sup>K<sup>−1</sup>. Its density is 19.3×10<sup>3</sup> kgm<sup>−3</sup>. Calculate the specific heat capacity of gold and the heat capacity per unit volume. What is the heat capacity of 4×10<sup>6</sup> kg of gold? (This is roughly the holdings of Fort Knox.)

1 mole of Au is $197$ g.

$$
\therefore c &= \frac{25.4 \text{ J mol}^{-1}\text{ K}^{-1}}{0.197\text{ kg mol}^{-1}} \cdot 19.3 \cdot 10^3 \text{ kg m}^{-3} \\
&= 2.5 \cdot 10^6 \text{ J m}^{-3} \text{ K}^{-1}
C &= \frac{25.4 \text{ J mol}^{-1}\text{ K}^{-1}}{0.197\text{ kg mol}^{-1}} \cdot 4 \cdot 10^6 \text{ kg} //
&= 5.2 \cdot 10^8 \text{ J K}^{-1}
$$
