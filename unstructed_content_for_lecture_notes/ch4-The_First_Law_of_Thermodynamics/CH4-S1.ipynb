### 4.3 Work done during volume changes

In thermodynamics, work is defined generally as energy transferred across a system boundary by any means other than heat. Expressions such as force multiplied by displacement arise as special cases of this definition when analysing mechanical systems, but it can be considered more generally as part of the thermodynamic framework.

Now lets take the concept of work and apply it to a thermodynamic system (any collection of objects or matter that is conveniently treated as a single unit for analysis, and that may exchange energy with its surroundings).

[FIGURE 19.3]

Consider a simple example of a thermodynamics system, a gas enclosed in a cylinder with a moveable piston. This system is coming in engines (e.g. internal combustion) and compressors (e.g. in refrigerators).

[FIGURE 19.4]

When a system such as a gas changes volume against a moveable piston, it can do work on its surroundings. An expanding gas pushes the piston outward and does positive work, whilst when the piston moves inward and the gas is compressed, work is done on the gas (this can also be thought of as the gas doing negative work). At the molecular level, work only occurs when gas molecules collide with a moving piston, not a stationary wall.

[FIGURE 19.5]

See the figure above, which shows a system (gas, liquid, solid) in a hollow cylinder with one end enclosed with a moveable piston. The piston has a cross-sectional surface are equivalent to the cylinder ($A$), and the enclosed gas exerts a pressure $p$ on the piston face. The total force exerted on the piston is $F=pA$. If in response to this force the piston moves an infinitesimal distance $dx$, the the work done in this process is:

$$
dW = F \ dx = pA \ dx.
\tag{4.11}
$$

Since $Adx = dV$, where dV is the infinitesimal volume change of the system, the work done by the system in this infinitesimal volume change is simply re-expressed as:

$$
dW = p \ dV.
\tag{4.12}
$$

In a finite change of volume, $V_1$ to $V_2$:

$$
W = \int_{V_1}^{V_2} p dV.
\tag{4.13}
$$

During a volume change, the system pressure generally varies, so evaluating Eq. (4.13) requires knowledge of the pressure-volume relationship $p(V)$. This work is represented graphically on a *pV-diagram* (see Chapter 3) as the area under the $p(V)$ curve between the points $V_1$ and $V_2$.

For **expansion** ($V_2 > V_1$, the area is positive and the system does positive work on its surroundings. For a **compression** ($V_2 < V_1$), the area is negative and the system does negative work. In the special case where pressure is constant throughout the whole process, Eq. (4.13) reduces to:

$$
W = p(V_2 - V_1).
\tag{4.14}
$$

For situations where the volume is constant, $dV = 0$ and hence no work is done.

### 4.4 Work and heat in a thermodynamic process

We now consider how the work performed by a system and the heat transferred to it during a thermodynamic process depend on the specific path and conditions under which the process occurs.

To understand intuitively how a path can influence the work performed by a system, picture how the difficulty rating of hiking trails which all lead to the same destination can challenge you differently, requiring more or less work on your part depending on the journey.

When a thermodynamic system described using pressure and volume ($p$, $V$), evolves from an initial state ($p_1$, $V_1$) to a final state ($p_2$, $V_2$), it passes through a sequence of intermediate equilibrium states called a **path**. Infinitely many paths can connect the same initial and final states on a pV-diagram, in the same manner in which two points on a graph can be connected via infinitely many curves.

    - Continuous variation of pressure with volume $\rightarrow$ the work differs from both piecewise paths preceding.

Like work, heat transfer in a thermodunamic process is path dependent. For example, consider an isothermal expansion (Section XYZ, constant temperature) of an ideal gas:

Initial state: $V_1=2.0 \ L$, $T=300 \ K$

Final state: $V_2=5.0 \ L$, $T=300 \ K$

### 4.5 The first law of thermodynamics

It may seem obvious to us today that heat and work represent forms of transient energy, but this understanding came about over a period of gradual discoveries. In the late eighteenth and early nineteenth centuries, heat was not initially regarded as a form of energy at all. Instead, it was treated as a material substance. The eventual recognition of heat and work as equivalent forms of energy was a major conceptual breakthrough that laid the foundation for the modern interpretation of thermodynamics.

**1789 - Antoine Lavoisier:**
Proposes the *caloric theory*, transferring heat as a conserved, weightless fluid responsible for thermal phenomena and combustion - (https://doi.org/10.1017/S000708740001195X).

**1798 - Benjamin Thompson (count Rumford):**
Demonstrates that friction during cannon boring can generate seemingly unlimited heat, contradicting the idea that caloric is a conserved substance - (https://doi.org/10.1098/rstl.1798.0006).

**1942 - Julius Robert Mayer:**
Proposes a quantitative relationship between mechanical work and heat, arguing that both are manifestations of a conserved quantity - energy - ().

**1840-1845 - James Prescott Joule:**
Performs precise experiments (including paddle-wheel and electrical heating) to measure the **mechanical equivalent of heat**, showing that a fixed amount of work always produces the same amount of heat - (https://doi.org/10.1098/rstl.1850.0004).

**Outcome:**
Heat came to be recognised as a form of energy rather than a material substance, leading to the abandonment of caloric theory and the formulation of the **first law of thermodynamics**, often summarised as:

$$
\text{"Energy is conserved and heat and work are both forms of energy."}
$$


#### Internal Energy

A thermodynamic system is characterised by its **internal energy**, denoted $U$. Internal energy represents the total energy associated with the system's internal degrees of freedom; this includes:

- The combined kinetic energy of the constituent particles (Translational, Rotational, Vibrational).
- The potential enegies arising from interactions between those particles.

The internal energy is a **state function**: its value depends only on the current state of the system, not on how that state was reached. Therefore, it contains no information about whether the energy encapsulated within the systen came from heat or work.


#### Heat and work

The internal energy of a system can only be changed by transferring energy across the system boundary. In thermodynamics, there are two distinct modes of energy transfer:

- **Heat** $Q$: energy transferred due to a temperature difference.
- **Work** $W$: energy transferred by all other mechanisms.

Heat and work are **not state functions**. Their values depend on the process by which the system evolves between states. Once a process is complete, the final state of the system contains no information about how much of the energy change arose from heat transfer and how much arose from work.

#### Sign Conventions and the First Law

The **first law of thermodynamics** expresses conservation of energy for a thermodynamic system by relating changes in internal energy to heat and work.

Different sign conventions are used in the literature to define work. To avoid confusion, we will state both conventions explicitly before selecting one to proceed with for the remainder of the course.

If work is defined as being **positive when work is done by the system**, $W = W_{by}$ and the first law may be written as:

$$
\Delta U = Q - W_{\text{by}},
\tag{4.15}
$$

where:
- $Q > 0$ corresponds to heat added to the system. If it's negative then heat is lost.
- $W_{\text{by}} > 0$ corresponds to work done by the system on its surroundings.

Alternatively, if work is defined as **positive when done on the system**, $W=W_{on}$, and the first law may be written as:

$$
\Delta U = Q + W_{\text{on}},
\tag{4.16}
$$

with
$$
W_{\text{on}} = -\, W_{\text{by}}.
$$

Both expressions describe the same physical principle - energy conservation - and differ only by the chosen sign convention.

> **Throughout this course, we adopt the convention that work is positive when it is done ON the system.**

Accordingly, we will write the first law using Eq. (4.16), with:
- $Q > 0$ when heat is added to the system,
- $Q < 0$ when heat is removed from the system,
- $W > 0$ when work is done on the system,
- $W < 0$ when work is done by the system.

When further clarification is warranted, we will expliticly make use of the labels: $Q_{added}$, $Q_{removed}$, $W_{on}$, $W_{by}$.

#### Cyclic Processes:

[FIGURE OF A CYCLIC PROCESS GRAPHED]

A **cyclic process** is a thermodynamic process in which a system evolves through a sequence of states and eventually returns to its initial state. On a pV-diagram, a cyclic process appears as a closed loop.

Because the system ends in the same state in which it began, all state variables return to their original values. In particular, the internal energy at the end of the cycle is equal to the internal energy at the beginning, so the resulting change in internal energy over a complete cycle is therefore zero.

$$
U_2 = U_1 \therefore \Delta U = U_2 - U_1 = 0.
$$

Notably, the temperature of the system also goes unchanged. Using the first law of thermodynamics:

$$
\Delta U = Q \ + \ W = Q_{added} + W_{on},
$$

we find that for a cyclic process,

$$
0 = Q \ + \ W,
$$

or equivalently,

$$
Q_{added} \ = \ - W_{on} \ = \ W_{by}
$$

This means that, over a complete cycle, the net heat added to the system is equal in magnitude and opposite in sign to the net work done on the system. Neither the heat nor work needs to be zero individually, only their sum must vanish.

The work appearing in the first law is defined as the work done **on** the system. However, on a pV-diagram, the mechanical work is given by the integral

$$
W_{by} = \int p dV,
\tag{4.17}
$$

which represents the work done by the gas. Therefore to make this suitable to our convention, we slightly amend Eq. (4.17) by multiplying the integral by -1, yielding:

$$
W_{on} = - \int p dV.
$$

With this expression, compression ($V_2 < V_1$) corresponds to positive $W$, and expansion  ($V_2 > V_1$) corresponds to negative $W$. When computing the integral of a cyclic process, the limits of integration form a closed loop and we write the work done on the gas as:

$$
W = - \oint p dV.
\tag{4.18}
$$

The sign of the integral itself depends on the direction in which the loop is traversed on the pV-diagram.

- A **clockwise cycle** corresponds to a positive value of $\oint p dV$, resulting in an overall negative value. In this case, work is done **by** the gas on its surroundings.

- A **counter-clockwise** cycle corresponds to a negative value of $\oint p dV$, hence (through a double-negative), the work is positive meaning work is done **on** the system.

Cyclic processes are fundamental to the operation of heat engines. In a heat engine, the system absorbs heat from a hot reservoir and performs net work on its surroundings over each cycle, while its internal energy remains unchanged.

Do note, in an isolated system, there is is no heat exchange carried out between said system and it's surroundings, as well as no work done on or by the surroundings, therefore $Q = W = 0$. Consequently, the internal energy remains constant - thus we conclude that the internal energy of an isolated system does not change during any process.

### 4.6 Thermodynamic Processes

Now let us consider various cases for gas expansion...

##### 1 - Adiabatic Process:

An **adiabatic** process is one in which there is no heat transfer into or out of the system (hence, $Q=0$). Therefore, in adiabatic processes the changes is temperature are brought about entirely through work, and not by heat transfer. Practically speaking, this can be achieved through sufficient system isolation; or by completing the process sufficiently quick such that heat does not have enough time to flow:

$$
\Delta U = W_{on} = - W_{by}\ \ \text{(adiabatic process)}
$$

During adiabatic expansion, the volume of the system is increased without allowing for heat transfer. Expansion means work done is being done by the system, which can be interpreted as negative work done on the system. Therefore the change in internal energy is negative and the temperature is typically also reduced.

During adiabatic compression, the volume of the system decreases without allowing for heat transfer. Compression means that the work done on the system is positive; hence the change in internal energy reflects this which typically corresponds to an increase in the system's temperature.

##### 2 - Isobaric Process:

An **isobaric** process is one in which the **pressure remains constant**. Such processes commonly occur when a gas is allowed to expand or contract against a constant external pressure; consider a gas beneath a freely moving piston, or boiling water exposed to atmospheric pressure.

On a p-V diagram, an isobaric process is represented by a horizontal line, since the pressure does not change whilst the volume increases or decreases.

- Initial state: ($p$, $V_i$)
- Final state: ($p$, $V_f$)

The area under this horizontal line corresponds to the work done by the gas. Simply multiply the difference between the final and initial volume by the constant pressure, and minus one:

$$
W = W_{on} = -p  (V_f - V_i) = - W_{by}
\tag{4.19}
$$

In an isobaric process, **none** of the quantities $\Delta U$, $Q$ or $W$ are necessarily zero. The first law still applies as normal. Unlike in the to-be-explained **isothermal** process, the temperature generally changes, so the internal energy also changes.

- During **isobaric expansion**, heat added to the gas is shared between: (1) increasing internal energy (raising temperature) and (2), doing work on the surroundings.
- During **isobaric compression**, work is done on the gas and heat must typically be removed.

##### 3 - Isothermal Process

An **isothermal** process is one which occurs at constant temperature, $\Delta T = 0$. For a process to remain isothermal, heat must flow into or out of the system slowly enough that thermal equilibrium with the surroundings is maintained at all times. In practice, this requires good thermal contact with a heat reservoir and quasi-static motion of the piston.

For many systems, internal energy depends on both temperature and volume. However, for an ideal gas, the internal energy depends only on temperature. Therefore in an isothermal process involving an ideal gas, $\Delta U = 0$. This is a special but extremely important case.

Starting from the first law:

$$
\Delta U = Q_{added} + W_{on}
$$

$$
0 = Q_{added} + W_{on}
$$

$$
Q_{added} = - W_{on} = W_{by}
$$

Therefore all the heat added to the system is converted directly into work done by the gas. No energy gets stored internally. Now let us consider a gas in a cylinder with a movable piston:

[DIAGRAM OF OF A CYLINDER WITH A MOVEABLE PISTON LID]

Initially, the gas has volume $V_1$ and temperature $T_1$. As the piston moves outward, the gas expands and does work on the surroundings. Expansion by itself would cool the gas. To keep the temperature constant, heat must flow into the gas from the surroundings.

(Reminder: a compressed gas tends to heat up and an expanding gas tends to cool down. This implies that heat must be added during isothermal expansion to maintain a constant temperature.)

So what is the work done by the gas? Let's consult a pV-diagram...

[GIVE IT UP FOR ANOTHER PV DIAGRAM]

The isothermal path follows a curve defined by $pV=nRT$ in which $T$ is constant. The pressure decreases smoothly as volume increases. The area under the curve represents the work done by the gas. Compared to an adiabatic expansion from the same initial point, the isothermal curve is less steep, because heat input prevents the temperature from dropping. The work done *by* the gas is given by:

$$
W_{by} = \int_i^f pdV,
$$

using the ideal gas law,

$$
p = \frac{nRT}{V}.
$$

We can sub this expression for $p$ into the integral:

$$
\begin{align}
W_{by} &= nRT \int_i^f \frac{1}{V} dV \\
&= nRT \ \left[ \ln(V) \right]_i^f \\
&= nRT \ \ln\left(\frac{V_f}{V_i} \right)
\end{align}
\tag{4.20}
$$

Where $W_{by} = Q_{added}$. *In summary, in an ideal isothermal system, the temperature remains constant ($\Delta T = 0$), the internal energy change is zero ($\Delta U = 0$), and the heat added is equal to the work done by the gas.*

```{admonition} Gas expansion and compression
:class: example

1 mol ($n=1$) of gas is kept at zero degrees celcius (273 K) during an expansion from 3 L to 10 L.

**(a) How much work is done by the gas? (Isothermal expansion step)**

$$
\begin{align}
W_{by} &= nRT ln\left(\frac{V_f}{V_i} \right) \\
&= 1 \times 8.31 \times 273 \times ln \left(\tfrac{10}{3}\right) \\
&= 2.7 \times 10^3 \ J
\end{align}
$$

**(b) How much heat is transferred into the system in the process?**

Remember that for isothermal processes, $Q_{added} = W_{by}$, so this quantity is also $2.7 \times 10^3 \ J$, as all of the heat is converted into work.

**(c) How much work is done to return the gas to its original volume via an Isobaric process?**

$$
\begin{align}
W_{by} &= p (V_f - V_i) \\
&= \frac{nRT}{V_i} (V_f - V_i) \\
&= \frac{1 \times 8.31 \times 273}{10\times 10^{-3}} (3\times 10^{-3} - 10\times 10^{-3}) \\
&= - 1.6\times 10^{3}
$$

The negative sign here indicates that work is being done *ON* the gas during compression.

#### Partial Pressure versus vapour pressure

It is important to understand the difference between vapour pressure and partial pressure. Partial pressure is the pressure of a gas if it occupies the entire volume.

For example, air at a pressure of 100 kPA is made up of 2% water vapour, what is the partial pressure of the water vapour?

$$
P_{water} = \left( \frac{V_{water}}{V_{total}} \right) p_{total} = 100 \ \text{kPa} \times 2\% = 2 \ \text{kPa}
$$

Vapour pressure is the pressure exerted by a vapour in thermal equilibrium with its condensed phase. It tells us how much vapour we can hold in a gas before we have to form a liquid.

- When partial pressure > vapour pressure $\rightarrow$ formation of liquid $\rightarrow$ clouds are formed.

- When partial pressure < vapour pressure $\rightarrow$ all the water stays in vapour form.

### 4.8 Internal Energy of an ideal gas: Free Expansion

In a free expansion, a gas expands into a vacuum inside an insulated container. No heat enters or leaves the system and the gas does no work: $Q = W = 0$ thus $\Delta U = 0$. Does the temperature of the gas change?

Suppose that the temperature *did* change during free expansion whilst the internal energy stayed constant. We would then have to conlcude that the internal energy depends not only on temperature $T$, but also on volume $V$ or pressure $p$. However, experiments show that for a low-density (ideal) gas, the temperature does not change during free expansion, despite the fact that $V$ and $p$ do. This leads us to a crucial conclusion:

> **For an ideal gas, the internal energy, $U$, depends only on temperature $T$, not on pressure or volume.**

This property, together with the ideal-gas equations of state is a fundamental part of the ideal-gas model and will be used repeatedly. For non-ideal gases, experiments do show a temperature change during free expansion despite an unchanging internal energy ($\Delta U = 0$). This means that internal energy cannot entirely depend on temperature; it must also depend on pressure and/or volume. Considering the microscopic reality of gases, this result is not surprising. Non-ideal gases have intermoleculer forces; as molecules move further apart from each other during expansion the experience an increase in their potential energies. If total internal energy remains constant the kinetic energy must drop accordingly - corresponding to a drop in temperature.

### 4.9 Heat capacitites for an ideal gas

To understand how heat affects a gas's internal energy, note that, in general, the internal energy depends on both temperature and volume - $U = U(V, T)$. Therefore a small change in internal energy can be expressed in terms of small changes in temperature and volume.

$$
dU = \left( \frac{\partial U}{\partial T} \right)_V dT + \left( \frac{\partial U}{\partial V} \right)_T dV.
\tag{4.22}
$$

Note that:

$$
dU = dQ + dW,
$$

and

$$
dW = - p dV.
$$

Therefore rearranging the first law using $dW = - pdV$ yields:

$$
dQ = dU + pdV,
$$

and now using eq. (4.22) we have that:

$$
dQ = \left( \frac{\partial U}{\partial T} \right)_V dT + \left[ \left( \frac{\partial U}{\partial V} \right)_T + p \right] dV.
\tag{4.23}
$$

We then divide eq. (4.23) by dT to obtain:

$$
\frac{dQ}{dT} = \left( \frac{\partial U}{\partial T} \right)_V + \left[ \left( \frac{\partial U}{\partial V} \right)_T + p \right] \frac{dV}{dT}.
$$

This relationship holds for any changes in temperature or volume, but our goal is to determine how much heat myst be added to change the temperature under specific constaints. The first case is when the volume is held constant, which leads to the definition of **heat capacity at constant volume**, $C_V$, which describes the heat required to raise the temperature of a system without changing its volume.

$$
C_V = \left(\frac{\partial Q}{\partial T} \right)_V = \left(\frac{\partial U}{\partial T} \right)_V
\tag{4.24}
$$

The **heat capacity at constant pressure** is then given by:

$$
C_p = \left(\frac{\partial Q}{\partial T} \right)_p = \left( \frac{\partial U}{\partial T} \right)_V + \left[ \left( \frac{\partial U}{\partial V} \right)_T + p \right] \left( \frac{dV}{dT} \right)_p,
\tag{4.25}
$$

such that:

$$
C_p - C_V = \left[ \left( \frac{\partial U}{\partial V} \right)_T + p \right] \left( \frac{dV}{dT} \right)_p.
\tag{4.26}
$$

Recall that the heat capacity has units of $\left[ \text{J} \cdot \text{K}^{-1} \right]$ and refers to a specific amount of gas. It is often practical to express heat capacity per mole or per unit mass; the lower is denoted using a lowercase 'c' and is referred to as the **specific heat capacity**.

$$
c_V = \frac{C_V}{M},
\tag{4.27}
$$

$$
c_p = \frac{C_p}{M},
\tag{4.28}
$$

where $M$ is the mass of the material. Specific heat capacities have units of $\left[ \text{J} \cdot \text{K}^{-1} \cdot \text{kg}^{-1} \right]$.

Let us now put this into practice for a monatomic gas. For an ideal monatomic gas, the internal energy arises solely from its constituents kinetic energy and equals $U = \tfrac{3}{2}RT$ per mole. Consequently, the internal energy depends only on the temperature. Hence:

$$
\left( \frac{\partial U}{\partial V} \right)_T = 0.
$$

The equation of state for 1 mole of an ideal gas is simply $pV = RT$. Rearranging to make $V$ the subject on the LHS and differentiating with respect to $T$ at constant pressure gives us:

$$
\left( \frac{\partial V}{\partial T} \right)_p = \frac{R}{p}.
$$

Supplying this result and its predecessor to eq. (4.26) we have that

$$
C_p - C_V = \left[ \left( \frac{\partial U}{\partial V} \right)_T + p \right] \left( \frac{dV}{dT} \right)_p = R.
$$

Because $U = \tfrac{3}{2}RT$, we therefore have that:

$$
C_V = \left( \frac{\partial U}{\partial T} \right)_V = \frac{3}{2}R \ \text{per mole},
\tag{4.29}
$$

and

$$
C_p = C_V + R = \frac{5}{2}R \ \text{per mole}.
\tag{4.30}
$$

The ratio of $C_p$ and $C_V$ is a very important and useful quantity as named the **adiabtatic index** ($\gamma$):

$$
\gamma = \frac{C_p}{C_V}.
\tag{4.31}
$$

For an ideal monatomic gas:

$$
\gamma = \frac{C_p}{C_V} = \frac{C_V + R}{C_V} = 1 + \frac{R}{C_V} = \frac{5}{3}.
$$

### 4.10 pV$^{\gamma}$ for adiabatic processes (Derivation for pV$^{\gamma}$ = constant)

For a perfect ideal gas, we have that $pV^{\gamma}$ is constant for all adiabatic processes (recall that adiabatic refers to contexts in which heat transfer ($Q$) is zero). This does not mean that each of the parameters $p$, $V$ and $\gamma$ are constant, we instead find that $\Delta p$, $\Delta V$ and $\Delta \gamma$ are all nonzero, but their resulting values combine to reform the constant $pV^{\gamma}$.

We will prove that $pV^{\gamma}$ is a constant. Starting with the internal energy of an ideal gas:

$$
U = \tfrac{3}{2}nRT,
$$

we differentiate both sides with respect to temperature ($T$), then multiply both sides by $dT$...

$$
\frac{dU}{dT} = \tfrac{3}{2}nR = C_V n \rightarrow dU = C_V n \ dT.
$$

Now we have an expresion for $dU$ which we can relate to the first law of thermodynamics, remembering that heat transfer for an adiabatic process is zero:

$$
dU = Q - W_{by} = 0 - pdV = - pdV,
$$

so we have

$$
C_V n \ dT = - pdV \rightarrow dT = \frac{pdV}{C_Vn}.
$$

Now that we've related these two expressions for internal energy, let us consider the differential form of the ideal gas law:

$$
pV = nRT,
$$

$$
Vdp + pdV = nRdT = - \frac{nRpdV}{C_Vn} = - \frac{RpdV}{C_V}.
$$

Divide both sides by pV and recall that $R = C_p - C_V$...

$$
\frac{Vdp}{pV} + \frac{pdV}{pV} = - \frac{RpdV}{C_VpV},
$$

$$
\frac{dp}{p} + \frac{dV}{V} = - \frac{R \ dV}{C_VV} = - \frac{(C_p - C_V)}{C_V} \frac{dV}{V}.
$$

Now, recall that:

$$
\gamma = 1 + \frac{R}{C_V} = 1 + \frac{C_p - C_V}{C_V};
$$

hence,

$$
\frac{dp}{p} + \frac{dV}{V} = - (\gamma - 1) \frac{dV}{V} = (1 - \gamma) \frac{dV}{V} = \frac{dV}{V} - \gamma \frac{dV}{V}.
$$

The $\tfrac{dV}{V}$ terms on either side cancel each other out through redundancy, and by adding the term on the RHS such that it equals zero, we are left with:

$$
\frac{dp}{p} + \gamma \frac{dV}{V} = 0.
$$

Now, we are left with incomplete integrands. Recognise that when we integrate the RHS that we are left with a constant (as a constant differentiated becomes zero):

$$
\int{\frac{1}{p}dp} + \gamma \int{\frac{1}{V}dV} = \text{constant},
$$

$$
\ln{(p)} + \gamma \ln{(V)} = \text{constant}.
$$

A sum of logs $A$ and $B$ is equivalent to a log of the product $AB$. Knowing this, and the fact that $a \times \log(B) = \log(B)^a$, we can rewrite the above equation as:

$$
\ln(pV^{\gamma}) = \text{constant}.
$$

Finally, to remove the logarithm from the LHS we exponentiate. However, since $e^{\text{constant}}$ is still a constant, we surmise that:

$$
pV^{\gamma} = \text{constant},
$$

and hence,

$$
p_1V_1^{\gamma} = p_2V_2^{\gamma} = \text{constant}
$$

for any adiabatic process.
