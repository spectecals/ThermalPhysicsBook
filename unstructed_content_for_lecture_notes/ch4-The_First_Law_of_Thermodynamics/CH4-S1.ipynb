### 4.3 Work done during volume changes

In thermodynamics, work is defined generally as energy transferred across a system boundary by any means other than heat. Expressions such as force multiplied by displacement arise as special cases of this definition when analysing mechanical systems, but it can be considered more generally as part of the thermodynamic framework.

Now lets take the concept of work and apply it to a thermodynamic system (any collection of objects or matter that is conveniently treated as a single unit for analysis, and that may exchange energy with its surroundings).

[FIGURE 19.3]

Consider a simple example of a thermodynamics system, a gas enclosed in a cylinder with a moveable piston. This system is coming in engines (e.g. internal combustion) and compressors (e.g. in refrigerators).

[FIGURE 19.4]

When a system such as a gas changes volume against a moveable piston, it can do work on its surroundings. An expanding gas pushes the piston outward and does positive work, whilst when the piston moves inward and the gas is compressed, work is done on the gas (this can also be thought of as the gas doing negative work). At the molecular level, work only occurs when gas molecules collide with a moving piston, not a stationary wall.

[FIGURE 19.5]

See the figure above, which shows a system (gas, liquid, solid) in a hollow cylinder with one end enclosed with a moveable piston. The piston has a cross-sectional surface are equivalent to the cylinder ($A$), and the enclosed gas exerts a pressure $p$ on the piston face. The total force exerted on the piston is $F=pA$. If in response to this force the piston moves an infinitesimal distance $dx$, the the work done in this process is:

$$
dW = F \ dx = pA \ dx.
\tag{4.11}
$$

Since $Adx = dV$, where dV is the infinitesimal volume change of the system, the work done by the system in this infinitesimal volume change is simply re-expressed as:

$$
dW = p \ dV.
\tag{4.12}
$$

In a finite change of volume, $V_1$ to $V_2$:

$$
W = \int_{V_1}^{V_2} p dV.
\tag{4.13}
$$

During a volume change, the system pressure generally varies, so evaluating Eq. (4.13) requires knowledge of the pressure-volume relationship $p(V)$. This work is represented graphically on a *pV-diagram* (see Chapter 3) as the area under the $p(V)$ curve between the points $V_1$ and $V_2$.

For **expansion** ($V_2 > V_1$, the area is positive and the system does positive work on its surroundings. For a **compression** ($V_2 < V_1$), the area is negative and the system does negative work. In the special case where pressure is constant throughout the whole process, Eq. (4.13) reduces to:

$$
W = p(V_2 - V_1).
\tag{4.14}
$$

For situations where the volume is constant, $dV = 0$ and hence no work is done.

### 4.4 Work and heat in a thermodynamic process

We now consider how the work performed by a system and the heat transferred to it during a thermodynamic process depend on the specific path and conditions under which the process occurs.

To understand intuitively how a path can influence the work performed by a system, picture how the difficulty rating of hiking trails which all lead to the same destination can challenge you differently, requiring more or less work on your part depending on the journey.

When a thermodynamic system described using pressure and volume ($p$, $V$), evolves from an initial state ($p_1$, $V_1$) to a final state ($p_2$, $V_2$), it passes through a sequence of intermediate equilibrium states called a **path**. Infinitely many paths can connect the same initial and final states on a pV-diagram, in the same manner in which two points on a graph can be connected via infinitely many curves.

    - 3 $\rightarrow$ 2: The pressure decreases however the volume stays constant $\rightarrow$ no work is done.

- Path 1 $\rightarrow$ 3 $\rightarrow$ 2

- Path 1 $\rightarrow$ 4 (smooth curve)
    - Continuous variation of pressure with volume $\rightarrow$ the work differs from both piecewise paths preceding.

Like work, heat transfer in a thermodunamic process is path dependent. For example, consider an isothermal expansion (Section XYZ, constant temperature) of an ideal gas:

Initial state: $V_1=2.0 \ L$, $T=300 \ K$

Final state: $V_2=5.0 \ L$, $T=300 \ K$

- **Path A (quasi-static isothermal expansion with a piston):** The gas expands slowly against a movable piston while heat is supplied to maintain constant temperature. The gas absorbs a definite, nonzero amount of heat and does work.

- **Path B (free expansion into a vacuum):** The gas rapidly expands rapidly into an evacuated region inside an isulated container. No work is done and no heat is transferred. For an ideal gas, the temperature remains unchanged, so the final state is identical as in the case of Path A.

Although the final and initial state are the same, the heat transferred and work done are different because the intermediate states (the path) differ.

Key Conclusion: Heat and work are path-dependent quantities, so they cannot be assigned definite values for a given thermodynamic state and it is meaningless to say that a system "contains" heat or work. In contrast, internal energy is a state function, uniquely determined by the system's state, which is why it plays a central and fundamental role in thermodynamics.

### 4.5 The first law of thermodynamics

##### TODO - Find the citation for Mayer's work.

It may seem obvious to us today that heat and work represent forms of energy, but this was not always the case. In fact, it took over 50 years to formulate this conceptualisation of the terms:

**1789 - Antoine Lavoisier:** Proposes the *caloric theory*, transferring heat as a conserved, weightless fluid that explains thermal phenomena and combustion - (https://doi.org/10.1017/S000708740001195X).

**1798 - Benjamin Thompson (count Rumford):** Demonstrates that friction during cannon boring can generate seemingly unlimited heat, contradicting the idea of caloric as a conserved substance - (https://doi.org/10.1098/rstl.1798.0006).

**1942 - Julius Robert Mayer:** Quantitatively links mechanical work and heat, proposing their equivalence and the conservation of energy - ().

**1840-1845 - James Prescott Joule:** Performs precise experiments (paddle wheel and electrical heating) to measure the **mechanical equivalent of heat**, showing that the same amount of work always produces the same amount of heat - (https://doi.org/10.1098/rstl.1850.0004).

**Outcome:**  Heat is recognised as a form of energy, leading to the abandonment of caloric theory and the formulation of the **first law of thermodynamics**, which was first formulated by Mayer and then Helmholtz as:

$$
\text{"Energy is conserved and heat and work are both forms of energy."}
$$

A system is characterised by its internal energy, $U$, defined as the total energy associated with its internal degrees of freedom. This includes the combined kinetic energies of all constituent particles comprising the system, as well as the potential energies arising from their mutual interactions.

The **internal energy** $U$ of a thermodynamic system characterises its state and can be changed by **heat transfer** or **work**. Heat $Q$ and work $W$ are **not state functions** because they depend on how energy is transferred; once the process is complete the system's state contains no information about whether the energy change came from heat or work.

By convention:

- $Q>0$ when heat is added to the system.

- $Q<0$ when heat is removed from the system.

- $W>0$ when work is done **on** the system.

- $W<0$ when work is done **by** the system.

- The change in internal energy between states $U_1$ and $U_2$ is $\Delta U - U_2 - U_1$.

If heat is added with no work done, $\Delta U=Q$; if work is done by the system with no heat transfer, $\Delta U=-W$. In general, when both occur we get the first law of thermodynamics:

$$
\Delta U = Q + W
\tag{4.15}
$$

#### Cyclic Processes:

[FIGURE OF A CYCLIC PROCESS GRAPHED]

A **cyclic process** is a thermodynamic process in which a system evolves through a sequence of states and eventually returns to its initial state. On a pV-diagram, a cyclic process appears as a closed loop. Because the final and initial states are identical, all state variables return to their original values. In particular, $U_{final} = U_{initial} \rightarrow \Delta U = 0$...

For a complete cycle, the start and end points must be the same, i.e. ($p_1$, $V_1$) = ($p_2$, $V_2$). Hence:

-> The temperature of the system is the same at the beginning and end of the process.
-> The internal energy at the end is also the same, hence $\Delta U = 0$.

Therefore, using the first law we can infer:

$$
\Delta U = Q \ + \ W
$$

$$
0 = Q_{added} \ + \ W
$$

$$
Q_{added} \ = \ - W
$$

Thus, in a cyclic process, the net heat added to the system equals the net work done **by** the system. Neither $Q$ nor $W$ need to be zero individually; only their sum must vanish. The net work done **by** the system over a cycle is given by the area enclosed by the loop on the corresponding pV-diagram:

- Clockwise cycle $\rightarrow$ positive work done by the system.
- Counterclockwise cycle $\rightarrow$ negative work (work done on the system).

This result underlies the operation of **heat engines**, which convert heat input into useful work through cyclic processes, even though the system's internal energy remains unchanged over each cycle.
