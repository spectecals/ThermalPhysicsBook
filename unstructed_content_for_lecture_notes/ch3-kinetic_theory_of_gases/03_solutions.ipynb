1. The theory of equipartition relates the internal energy of a gas to its temperature.


   1. What assumptions are made in the theory of equipartition? (2)

        <RAW HTML BREAK>
        <br><br>
        </RAW HTML BREAK>

      - Energy is evenly distributed across each DOF, aka thermal equilibrium
      - Classical mechanics applies, can sum contribution from each DOF
      - Intermolecular interactions negligible (ideal gas)
      - Atoms are point particles (have no size)

        <RAW HTML BREAK>
        <br><br>
        </RAW HTML BREAK>

   2. Which of these assumptions must be modified for real gases, and why? (2)

        <RAW HTML BREAK>
        <br><br>
        </RAW HTML BREAK>

      - Interactions are not negligible (van der Waals), alters force and therefore pressure
      - Particles have volume, therefore rates of collision and effective volume change

        <RAW HTML BREAK>
        <br><br>
        </RAW HTML BREAK>

   3. Explain the implications of the theory of equipartition for monatomic versus diatomic molecules. (2)

        <RAW HTML BREAK>
        <br><br>
        </RAW HTML BREAK>

      $$ U = \frac{\text{DOF}}{2} nRT $$

      - Monatomic gases: $\text{DOF}=3 \Rightarrow U=\frac{3}{2}nRT$
      - Diatomic gases (with vibrational modes frozen): $\text{DOF}=5 \Rightarrow U=\frac{5}{2}nRT$

        <RAW HTML BREAK>
        <br><br>
        </RAW HTML BREAK>

   4. State the internal energy of:
      - one mole of a monatomic gas, and  
      - one mole of a diatomic gas,  
      assuming a temperature of 300 K. (2)

        <RAW HTML BREAK>
        <br><br>
        </RAW HTML BREAK>

      $$RT = (8.314)(300) = 2.494 \times 10^{3}\ \text{J}

        U_{\text{mono}} = \frac{3}{2}RT = 3.74 \times 10^{3}\ \text{J}

        U_{\text{di}}   = \frac{5}{2}RT = 6.24 \times 10^{3}\ \text{J} $$

        <RAW HTML BREAK>
        <br><br>
        </RAW HTML BREAK>

   5. What are the implications of the theory of equipartition for gases at low and high temperatures? (2)

        <RAW HTML BREAK>
        <br><br>
        </RAW HTML BREAK>

      - High $T$: Z stays near 1 but additional degrees of freedom become active
      - Low $T$: intermolecular effects are stronger, Z correlates with pressure

2. The pressure $p$ in an ideal gas can be shown to be
   $p=\frac{2}{3}N_V\braket{E_k}$,
   where $N_V$ is the number of molecules per unit volume and $\braket{E_k}$ is the
   average kinetic energy of a molecule. By relating this expression to the pressure
   obtained from the ideal gas law, what can be concluded about the energy per degree
   of freedom in an ideal gas? (5)

    <RAW HTML BREAK>
    <br><br>
    </RAW HTML BREAK>

      Ideal gas law (molecular form):

      $$ p = N_V kT $$

      Equating expressions:

      $$N_V kT = \frac{2}{3} N_V \braket{E_k} \\
        \braket{E_k} = \frac{3}{2} kT $$

      Energy per quadratic degree of freedom:

      $$\frac{1}{2}kT$$

    <br><br>

3. Consider a gas that is raised from temperature $T_C$ to $T_H$. This process is
   conducted either at constant pressure or at constant volume. By considering these
   two processes, show that the ratio of the molar specific heat capacities at constant
   pressure $c_p$ and constant volume $c_V$ of an ideal gas is equal to $5/3$. (5)

    <RAW HTML BREAK>
    <br><br>
    </RAW HTML BREAK>

      $$ c_V \equiv \frac{1}{n}\left(\frac{dQ}{dT}\right)_V $$ 

      For a monatomic ideal gas:

      $$ U(T) = \frac{3}{2}nRT \rightarrow \frac{dU}{dT} = \frac{3}{2}nR. $$

      Constant volume:

      $$ \left(\frac{dQ}{dT}\right)_V = \frac{dU}{dT} = \frac{3}{2}nR \\
        c_V = \frac{1}{n}\frac{dQ}{dT}=\frac{3}{2}R $$

      Constant pressure:

      $$ c_P \equiv \frac{1}{n}\left(\frac{dQ}{dT}\right)_P \\
        dQ_P = dU + p\,dV $$
    
      Ideal gas law:

      $$ pV = nRT \\
        p\,dV = nR\,dT $$

      Insert $dQ_P$:

      $$ \frac{dQ_P}{dT} = \frac{3}{2}nR + nR = \frac{5}{2}nR $$
      
      Ratio:

      $$ \frac{c_P}{c_V} = \frac{5}{3} $$
