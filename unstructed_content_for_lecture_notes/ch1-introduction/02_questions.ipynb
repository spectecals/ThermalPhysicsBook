## Questions

1. What is the mass of 3 moles of carbon dioxide (CO<sub>2</sub>)? (1 mole of oxygen atoms has a mass of 16 g.)

2. Are there more water *molecules* in your body (assuming you are 100% water) or more water *droplets* in all the oceans in the world? There is approximately $10^{21}$ kilograms of water in all of the world's oceans. Use an estimate of the size of a typical droplet of water.

3. A system contains $n$ atoms, each of which can only have zero or one quanta of energy. How many ways can you arrange $r$ quanta of energy when:
    1. $n=20,\ r=10$;
    2. $n=2*10^{23},\ r=10^{23}$?
