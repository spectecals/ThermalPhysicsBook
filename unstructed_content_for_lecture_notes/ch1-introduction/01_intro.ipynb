## What is a mole?

In thermal physics, we deal with enormous numbers of atoms and molecules. These large numbers make it impossible to track individual particles, so we describe their behaviour statistically, in terms of averages. Even numbers we think of as huge, like the Earth’s population (~$10^9$) or the U.S. national debt (~$10^{13}$), are tiny compared with the numbers of atoms in ordinary matter (around $10^{23}$–$10^{25}$ per sample). For instance, 1 kg of nitrogen gas contains about 2 × $10^{25}$ molecules. A fast computer counting one molecule per clock cycle would take hundreds of millions of years just to count them! To handle such vast quantities, we use the concept of the mole, a convenient counting unit similar to a “dozen,” but much larger.

```{admonition} Definition:
:class: definition
A mole is the quantity of matter that contains as many objects (atoms, molecules, ions) as there are atoms in *exactly* 12 grams of $^{12}$C.
```
This number, called Avogadro’s number, is $N_A=6.022×10^{23}$. Thus, one mole of carbon, water, or sodium chloride each contains $6.022×10^{23}$ of their respective particles.

  - $k_B=1.38*10^{-23}\text{JK}^{-1}$

## Intrinsic vs extrinsic variables

### Intrinsic variables

These are variables where if you were to take a sample of your system, the values will not change. Notably these include:

- $P$, pressure
- $T$, temperature
- $\rho$, density
- $c_v$, heat capacity

Consider for each of these what would happen if you had a gas in a box, measured these values, and then repeated the process after taking a sample from the box. They will not change, whereas something like the volume clearly would.

### Extrinsic variables

These are variables which depend on the *amount of matter* in the system. Consider:

- $N$, number of atoms
- $V$, volume
- $U$, internal energy
- $Q$, heat

If you drank half a cup of warm coffee, you'll obviously find that the number of atoms, volume, internal energy and the heat it radiates all halve.

## The thermodynamic limit

Thermodynamics is a phenomenological theory of macroscopic variables and therefore only strictly holds true for large systems. In fact, we must begin by assuming that the number of particles, $N$, tends to infinity, as does the volume $V$, in such a way that the density $\rho$ of the gas remains constant and uniform,

$$N \rightarrow \inf, \qquad V \rightarrow \inf, \qquad \rho = \frac{N}{V}. $$



## Ideal Gases

This section introduces a very important concept called the *ideal gas law*, based around the understanding that the pressure of a gas depends on its temperature and volume.

- **Boyle's law** (who boiled a viper in a vacuum): Experiments show that, at constant temperature, the product of $P V$ remains constant,

$$ P \propto \frac{1}{V}. $$

- **Charles' law** (also an avid balloonist): When the pressure is held constant, the volume of a gas increases proportionally with temperature,

$$ V \propto T. $$

- **Gay-Lussac's law** (who discovered the composition of water): At constant volume, the pressure is proportional to temperature,

$$ P \propto T. $$

Combining these empirical results, alongside $\rho = \frac{N}{V}$, we find that $P V \propto N T$. Their proportionality is the aforementioned Boltzmann constant, which relates temperature to energy, and forms the **microscopic ideal gas equation**,

$$ P V = N k_B T \qquad \text{(microscopic)}. $$

Alternatively, we can define it macroscopically in terms of the numbers of moles, $n$, through $N=n * N_A$ and the universal gas constant, $R = N_A*k_B$, forming the **macroscopic ideal gas equation**,

$$ P V = n R T \qquad \text{(macroscopic)}. $$

This *ideal gas* model assumes that molecules:
1. move randomly,
2. collide elastically,
3. do not have intermolecular interactions (such as attracting or repelling each other) and
4. are point-like with negligible volume.

Under these assumptions, the gas pressure arises from molecules striking the container walls. Despite its simplicity, the model accurates describes many real gases under conditions of low pressure and high temperature. At higher densities or lower temperatures, deviations occur because real molecules have finite size and experience intermolecular forces.

## Combinatorial problems

Thermal physics studies how large collections of particles behave *statistically*. Since it is impossible to track every particle individually, we instead count how many distinct configurations a system can have that correspond to a given macroscopic state. This kind of counting is a combinatorial problem specifically important for indistinguishable particles.

### The Combination Formula

When counting configurations, we need to know how many ways we can choose a certain number of items from a larger set, **without regard to order**. The number of such combinations is given by:

$$ \frac{n}{k} = \frac{n!}{k!(n-k)!} $$

Here $n$ is the total number of available positions and $k$ the number of elements we choose to occupy them. $n!$ refers to "$n$ factorial", meaning $n * (n - 1) * (n - 2) * ... * 1$. This formula tells us how many distinct configurations are possible when order does not matter.
