**1.1 What is a mole?**

In thermal physics, we deal with enormous numbers of atoms and molecules. These large numbers make it impossible to track individual particles, so we describe their behaviour statistically, in terms of averages. Even numbers we think of as huge, like the Earth’s population (~$10^9$) or the U.S. national debt (~$10^{13}$), are tiny compared with the numbers of atoms in ordinary matter (around $10^{23}$–$10^{25}$ per sample). For instance, 1 kg of nitrogen gas contains about 2 × $10^{25}$ molecules. A fast computer counting one molecule per clock cycle would take hundreds of millions of years just to count them! To handle such vast quantities, we use the concept of the mole, a convenient counting unit similar to a “dozen,” but much larger.

```{admonition} Definition:
:class: note
A mole is the quantity of matter that contains as many objects (atoms, molecules, ions) as there are atoms in *exactly* 12 grams of $^{12}$C.
```
This number, called Avogadro’s number, is $N_A=6.022×10^{23}$. Thus, one mole of carbon, water, or sodium chloride each contains $6.022×10^{23}$ of their respective particles.

**1.2 Useful constants and variables**

Constants:

Other than Avogadro's number, other useful constants include:

- The Boltzmann constant, $k_B$
  - $k_B=1.38*10^{-23}JK^{-1}$
  - This is a constant that relates the average energy of a particle ina system to its temperatures. It serves as a conversion factor between temperature and energy.
- The universal gas constant, $R$
    - $R = 8.31 JK^{-1} \text{mol}^{-1}$
    - Like the Boltzmann constant, this is a proportionality constant relating the energy [J] and temperature [K] and the amount of substance used [mol]. In other words, it’s the molar equivalent to the Boltzmann constant: $R = NA*K_b$. This is the constant of proportionality used in the ideal gas law (Section XYZ)

Variables:

- Pressure, $P$
    - Pressure is a state variable law (Section XYZ) , definesd as the amount of <i>force (F) per unit area (A)</i>
    - $P = F/A [N/m^2]$
    - There is a zoo of units used to express pressure, which can be confusing for beginners since different fields favour different conventions. Below are commonly used units that you may encounter:
        - SI units: 
            - Pascal (Pa), 1 Pa = 1 $N/m^2$. This is the official unit for pressure, and the one we will use in this course.
        - Non-SI units: 
            - atmosphere (atm), 1 atm = 101,325 Pa
            - bars, 1 bar = 100,000 Pa
                - (notably very close to 1 atm)
            - pound per square inch (psi), 1 psi = 6,895 Pa
            - millimetres of mercury (Torr), 1 mmHg (or 1 Torr, slightly different) = 1/760 atm = 101,325/760 Pa = 133.322368421 Pa
