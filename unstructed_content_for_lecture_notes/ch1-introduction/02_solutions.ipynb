## Solutions

1. What is the mass of 3 moles of carbon dioxide (CO<sub>2</sub>)? (1 mole of oxygen atoms has a mass of 16 g.)

CO<sub>2</sub> has atomic numbers $Z = 12, 16$. The definition of the mole is the number of atoms required such that our carbon-12 weighs 12g, therefore we don't need to use Avagadro's number or find the weight of a singular molecule. 

$$ \text{sum of atomic numbers} * \text{number of moles} &= \text{mass of gas in grams} \\
(12 + 16*2) * 3 &= 44 * 3 = 132\ \text{g} $$

2. A typical bacterium has a mass of 10<sup>âˆ’12</sup> g. Calculate the mass of a mole of bacteria. (Interestingly, this is about the total number of bacteria living in the guts of all humans resident on planet Earth.) Give your answer in units of elephant-masses (elephants have a mass ~5000 kg).

Here we do require Avagadro's number, $6.022 * 10 ^(23)$, and it will help to write the elephant masses in the form of $5 * 10^3\  \text{kg}=5 * 10^6\ \text{g} = 1\ \text{g}/ 2*10^{-7}$.

$$
\text{single bacterium mass} * \text{Avagadro's number} / \text{elephant mass} = \text{elephants composed of bacteria} \\
10^{-12}\ \text{g} * 6.022*10^{23}\ \text{mol}^-1 * 2*10^{-7}\ \text{g}^-1 &= 2 * 6.022 * 10^{23 - 19} \\
&= 12.044 * 10^{4} = 120,000\ \text{elephants}.
$$

3. Are there more water *molecules* in your body (assuming you are 100% water) or more water *droplets* in all the oceans in the world? There is approximately $10^{21}$ kg of water in all of the world's oceans. Use an estimate of the size of a typical droplet of water.

Average weight of a European is $ \approx 70kg $. H<sub>2</sub>O has a molar mass of $2 + 16$ g. We can start by calculating how many moles of water would be required to weigh 70 kg, and then how many molecules total.

$$
70\ \text{kg} / 18\ \ text{g} \approx 3,900\ \text{moles} = 3,900\ \text{moles}* 6.022 * 10 ^ {23}\ \text{moles}^{-1} = 2.35 * 10^{27}\ \text{molecules}.
$$

Raindrops range from 0.5-5 mm typically, so we'll go with 1 mm diameter droplets. A 0.5 mm radius sphere has volume 
$$ 4/3 \pi (0.5 * 10^{-3})^3\ \text{m}^3 \approx 0.5 * 10^{-9}\ \text{m}^3 \\
1000 \text{kg water} == 1\ \text{m}^3\ \text{water} \\
0.5 * 10^{-9}\ \text{m}^3 = 0.5 * 10^{-6}\ \text{kg} \\
10^{21} * 2 * 10^6 = 2 * 10^{27} < 2.35 * 10^{27} $$

So, more droplets.
