## Solutions

1. What is the mass of 3 moles of carbon dioxide (CO<sub>2</sub>)? (1 mole of oxygen atoms has a mass of 16 g.)

CO<sub>2</sub> has atomic numbers $Z = 12, 16$. The definition of the mole is the number of atoms required such that our carbon-12 weighs 12g, therefore we don't need to use Avagadro's number or find the weight of a singular molecule. 

$$ \text{sum of atomic numbers} \cdot \text{number of moles} &= \text{mass of gas in grams} \\
(12 + 16\cdot 2) \cdot 3 &= 44 \cdot 3 = 132\ \text{g} $$

2. A typical bacterium has a mass of 10<sup>âˆ’12</sup> g. Calculate the mass of a mole of bacteria. (Interestingly, this is about the total number of bacteria living in the guts of all humans resident on planet Earth.) Give your answer in units of elephant-masses (elephants have a mass ~5000 kg).

Here we do require Avagadro's number, $6.022 \cdot 10 ^{23}$, and it will help to write the elephant masses in the form of $5 \cdot 10^3\  \text{kg}=5 \cdot 10^6\ \text{g} = 1\ \text{g}/ 2\cdot 10^{-7}$.

$$
\text{single bacterium mass} \cdot \text{Avagadro's number} / \text{elephant mass} &= \text{elephants composed of bacteria} \\
10^{-12}\ \text{g} \cdot 6.0210^{23}\ \text{mol}^{-1} \cdot 2\cdot 10^{-7}\ \text{g}^{-1} &= 2 \cdot 6.022 \cdot 10^{(23 - 19)} \\
&= 12.044 \cdot 10^{4} = 120,000\ \text{elephants}.
$$

3. Are there more water *molecules* in your body (assuming you are 100% water) or more water *droplets* in all the oceans in the world? There is approximately $10^{21}$ kg of water in all of the world's oceans. Use an estimate of the size of a typical droplet of water.

Average weight of a European is $ \approx 70kg $. H<sub>2</sub>O has a molar mass of $(2 + 16)$ g. We can start to solve this question by calculating how many moles of water would be required to weigh 70 kg, and then how many molecules total.

$$
70000\ \text{g} / 18\ \text{g mol}^{-1} &\approx 3900\ \text{mol} \\
3900\ \text{mol}\cdot 6.022 \cdot 10 ^ {23}\ \text{molecules per mol} &= 2.35 \cdot 10^{27}\ \text{molecules}.
$$

Raindrops range from 0.5-5 mm typically, so we'll go with 1 mm diameter droplets. A 0.5 mm radius sphere has volume

$$ 4/3 \pi (0.5 \cdot 10^{-3})^3\ \text{m}^3 &\approx 0.5 \cdot 10^{-9}\ \text{m}^3 \\
1000 \text{kg water} &\equiv 1\ \text{m}^3\ \text{water} \\
0.5 \cdot 10^{-9}\ \text{m}^3 &= 0.5 \cdot 10^{-6}\ \text{kg} \\
10^{21} \cdot 2 \cdot 10^6 &= 2 \cdot 10^{27} < 2.35 \cdot 10^{27} $$

Depending on your approximation of the droplet size, it could go either way! If you used 1 mm **radius** droplets, you'd have $ 8 \cdot 2 \cdot 10^{27} > 2.35 \cdot 10^{27} $.

4. A system contains $n$ atoms, each of which can only have zero or one quanta of energy. How many ways can you arrange $r$ quanta of energy when:
    1. $n=20,\ r=10$;
    2. $n=2\cdot 10^{23},\ r=10^{23}$?

Since atoms are indistinguishable, we'll need to use $^nC_R$, not $^nP_R$.

1. $n=20,\ r=10$;

$$ \frac{20!}{(20-10)! \cdot 10!} &= \frac{20!}{10!\cdot 10!} \\
&= 184756 
$$

2. $n=2\cdot 10^{23},\ r=10^{23}$

$$ \frac{2\cdot 10^{23}!}{10^{23}!(2\cdot 10^{23}-10^{23})!} = \frac{2\cdot 10^{23}!}{10^{23}!\cdot 10^{23}!} $$

This number is unfathomably huge. We're starting to approach the regime where we might need to use [tetration](https://en.wikipedia.org/wiki/Large_numbers#Standardized_system_of_writing) just in order to represent it. Doing math in this sort of regime is simply not feasible, no computer can perform this operation. It would require over $10^{22}$ digits which is just about manageable if we were able to use all the digital storage on the planet (the International Data Corporation's estimate of the total digital storage that will be manufactured in 2025 is only 22 ZB, or $22 \cdot 10^{21}$, so we'll need to borrow some more from the previous years too). The *only* way to approach this question is to find some approximate solution.

Luckily, there is one such approximation specifically for factorials, [Stirling's](https://en.wikipedia.org/wiki/Stirling%27s_approximation). 

$$ n! &\approx \sqrt{2\pi n}\frac{n}{e}^n \\
\text{ln } n! &\approx n \text{ln } n - n + O(\text{ln } n). $$

This approximation fails for $n!$ when $n=0$, but all subsequent $n$ is at least close, and for large $n$ the approximation is effectively exact. The error associated in this approximation is somewhere between $e^{\frac{1}{12n+1}} < \frac{n!}{S(n)} < e^\frac{1}{12n}$. For $n=10^{23}$, the relative error is on the order of $10^{-25}$, i.e. the *absolute* error will have $10^{22} - 25$ digits. This is still clearly unbelievably huge but in real life you're lucky if you can say you know the first 25 digits of a number with certainty.

We will omit the $O(\text{ln } n)$ term as we are only interested in the scale of the number and don't require digit-level accuracy. Let $n=10^23$,

$$
\text{ln } 2n! &\approx 2n \text{ln } 2n - 2n \\
\text{ln } (n! \cdot n!) &\approx 2 \text{ln } n! \\
&\approx 2(n \text{ln } n - n) \\
&= 2n \text{ln } n - 2n \\
\text{ln }\frac{2n!}{n! \cdot n!} &=  2n \text{ln } 2n - 2n - (2n \text{ln } n - 2n) \\
&= 2n(\text{ln } 2n - \text{ln } n) \\
&= 2n(\text{ln } \frac{2n}{n}) \\
&= 2n(\text{ln } 2) \\
&= 2 \cdot 10^{23} \text{ln } 2 \\
&= 1.39 \cdot 10^{23}
$$
