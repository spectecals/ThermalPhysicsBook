# Probability Theory - Questions

## Question 1

Suppose that you are offered a choice of the following: a certain £10 or the chance of rolling a fair die and receiving £36 if it comes up 5 or 6, but nothing otherwise. What is the economical choice to make and why?

---

## Question 2

Even though you have no symptoms, your doctor wishes to test you for a rare disease that only 1 in 10,000 people of your age contract. The test is 98% accurate, which means that if you have the disease, 98% of the times the test will come out positive, and 2% negative.

We will also assume that if you do not have the disease, the test will come out negative 98% of the time and positive 2% of the time. You take the test and it comes out positive. What is the probability that you have the disease?

---

## Question 3

Imagine that you have a sack with two balls in it that can be either red or green. There are three hypotheses for the distribution of colours for the balls:

1. both balls are red,
2. one ball is red and one ball is green,
3. both balls are green.

Initially, you have no information about which hypothesis is correct, and thus you assume that they are equally probable.

Suppose that you pick one ball out of the sack and it is green. What are the updated probabilities for each hypothesis now?

## Question 4

For a coin with the probability of getting heads \(p_{h} = 0.3\), what is the probability that in five tosses of a coin there will be two heads? Sketch the full binomial distribution for this coin being tossed five times.

What is the mean number of heads and corresponding standard deviation if the coin is thrown 10,000 times?
